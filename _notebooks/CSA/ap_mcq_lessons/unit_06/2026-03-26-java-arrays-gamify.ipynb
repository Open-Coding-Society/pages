## Learning Objectives

> **8.1** Represent collections of related object reference data using one-dimensional (1D) array objects.  
> **8.2** Traverse the elements in a 1D array.  
> **9.1** Represent collections of related object reference data using two-dimensional (2D) array objects.

## 1 — Array Declaration & Initialization

An **array** is a fixed-size, ordered collection of values of the same type.

```java
// Declare and allocate — elements default to 0 (int), false (boolean), null (Object)
int[] scores = new int[5];

// Declare with an initializer list (size is inferred)
int[] primes = {2, 3, 5, 7, 11};

// Object array — elements default to null
String[] names = new String[3];
```

| Type | Default value |
|------|---------------|
| `int`, `long`, `short`, `byte` | `0` |
| `double`, `float` | `0.0` |
| `boolean` | `false` |
| Object (`String`, etc.) | `null` |

## 2 — Indexing & `arr.length`

Java arrays are **0-indexed**: the first element is at index `0`, the last is at index `arr.length - 1`.

```java
int[] arr = {10, 20, 30, 40};

System.out.println(arr[0]);          // 10  ← first
System.out.println(arr[3]);          // 40  ← last
System.out.println(arr.length);      // 4

// arr[arr.length] → ArrayIndexOutOfBoundsException!
```

> **Tip:** `arr.length` is a *field*, not a method — no parentheses!

## 3 — Standard `for` Loop Traversal

Use an index variable when you need the position or want to modify elements.

```java
int[] nums = {1, 2, 3, 4, 5};

// Forward traversal
int sum = 0;
for (int i = 0; i < nums.length; i++) {
    sum += nums[i];
}
System.out.println(sum); // 15

// Backward traversal
for (int i = nums.length - 1; i >= 0; i--) {
    System.out.print(nums[i] + " "); // 5 4 3 2 1
}

// Every other element
for (int i = 0; i < nums.length; i += 2) {
    System.out.print(nums[i] + " "); // 1 3 5
}
```

## 4 — Enhanced `for` Loop (For-Each)

When you only need to **read** each element (no index, no modification), the enhanced `for` is cleaner.

```java
int[] scores = {85, 92, 78, 90};

// Read all elements
for (int score : scores) {
    System.out.println(score);
}

// ⚠️ This does NOT change the array:
for (int score : scores) {
    score = 100;  // 'score' is a copy — 'scores' is unchanged!
}
```

| | Standard `for` | Enhanced `for` |
|-|---------------|----------------|
| Access index | ✅ | ❌ |
| Modify elements | ✅ | ❌ |
| Read-only traversal | ✅ | ✅ (preferred) |

## 5 — 2D Arrays

A 2D array is an array of arrays — think of it as a grid of rows and columns.

```java
// Declare a 3-row × 4-col grid
int[][] grid = new int[3][4];

// Initializer list
int[][] matrix = {
    {1, 2, 3},
    {4, 5, 6},
    {7, 8, 9}
};

System.out.println(matrix[1][2]); // 6  (row 1, col 2)
System.out.println(matrix.length);        // 3  (rows)
System.out.println(matrix[0].length);     // 3  (columns)

// Nested traversal
for (int r = 0; r < matrix.length; r++) {
    for (int c = 0; c < matrix[r].length; c++) {
        System.out.print(matrix[r][c] + " ");
    }
    System.out.println();
}
```

## 6 — `java.util.Arrays` API

The `Arrays` class provides static utility methods for common array operations.

```java
import java.util.Arrays;

int[] arr = {5, 3, 8, 1, 9};

// Sort ascending (in-place)
Arrays.sort(arr);                       // {1, 3, 5, 8, 9}

// Readable string representation
System.out.println(Arrays.toString(arr)); // [1, 3, 5, 8, 9]

// Fill every element with a value
Arrays.fill(arr, 7);                    // {7, 7, 7, 7, 7}

// Copy first 3 elements into a new array
int[] copy = Arrays.copyOf(arr, 3);     // {7, 7, 7}

// Binary search (array must be sorted first!)
int idx = Arrays.binarySearch(arr, 5);  // returns index of 5
```

> **Remember:** `Arrays.sort()` modifies the original array. Use `Arrays.copyOf()` first if you need to keep the original.
