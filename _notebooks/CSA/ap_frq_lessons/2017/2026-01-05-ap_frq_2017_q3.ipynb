public int findLastOccurrence(String str) {
    int occurrence = 1;
    int index = findNthOccurrence(str, occurrence);
    
    while (index != -1) {
        occurrence++;
        int nextIndex = findNthOccurrence(str, occurrence);
        if (nextIndex == -1) {
            return index;
        }
        index = nextIndex;
    }
    
    return -1;
}
