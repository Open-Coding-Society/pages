## Part (a)

Write the WordMatch method scoreGuess. To determine the score to be returned, scoreGuess finds the number of times that guess occurs as a substring of secret and then multiplies that number by the square of the length of guess. Occurrences of guess may overlap within secret.

Assume that the length of guess is less than or equal to the length of secret and that guess is not an empty string.

````java
// CODE_RUNNER: WordMatch Part (a) Solution and Testing

// SubstringTester class

public class Main {
    /** The secret string. */
    private String secret;
    
    /** Constructs a WordMatch object with the given secret string of lowercase letters. */
    public Main(String word) {
        secret = word;
    }
    
    // ==================== PART (a) SOLUTION ====================
    /** Returns a score for guess, as described in part (a).
     * Precondition: 0 < guess.length() <= secret.length()
     */
    public int scoreGuess(String guess) {
        int count = 0;
        int index = 0;
        
        // Count all occurrences of guess in secret (including overlapping)
        while (index <= secret.length() - guess.length()) {
            if (secret.substring(index, index + guess.length()).equals(guess)) {
                count++;
            }
            index++;
        }
        
        // Calculate score: count × (length of guess)²
        return count * guess.length() * guess.length();
    }
    
    // ==================== TESTING PART (a) ONLY ====================
    public static void main(String[] args) {
        System.out.println("========================================");
        System.out.println("PART (a): scoreGuess Method");
        System.out.println("========================================\n");
        
        // Example 1: WordMatch game = new WordMatch("mississippi");
        System.out.println("Example 1: WordMatch game = new WordMatch(\"mississippi\");\n");
        Main game1 = new Main("mississippi");
        
        System.out.println("Testing with secret = \"mississippi\":");
        testScoreGuess(game1, "i", 4, "4 occurrences × 1² = 4");
        testScoreGuess(game1, "iss", 18, "2 occurrences × 3² = 18");
        testScoreGuess(game1, "issipp", 36, "1 occurrence × 6² = 36");
        testScoreGuess(game1, "mississippi", 121, "1 occurrence × 11² = 121");
        
        // Example 2: WordMatch game = new WordMatch("aaaabb");
        System.out.println("\n\nExample 2: WordMatch game = new WordMatch(\"aaaabb\");\n");
        Main game2 = new Main("aaaabb");
        
        System.out.println("Testing with secret = \"aaaabb\":");
        testScoreGuess(game2, "a", 4, "4 occurrences × 1² = 4");
        testScoreGuess(game2, "aa", 12, "3 occurrences × 2² = 12");
        testScoreGuess(game2, "aaa", 18, "2 occurrences × 3² = 18");
        testScoreGuess(game2, "aabb", 16, "1 occurrence × 4² = 16");
        testScoreGuess(game2, "c", 0, "0 occurrences × 1² = 0");
        
        System.out.println("\n========================================");
        System.out.println("Part (a) Testing Complete!");
        System.out.println("========================================");
    }
    
    // Helper method for testing
    private static void testScoreGuess(Main game, String guess, int expected, String calculation) {
        int actual = game.scoreGuess(guess);
        String status = (actual == expected) ? "✓ PASS" : "✗ FAIL";
        System.out.printf("  %s scoreGuess(\"%s\") = %d (expected: %d)%n", 
                         status, guess, actual, expected);
        System.out.printf("     Calculation: %s%n", calculation);
    }
}


Main.main(null);
````

````java
// CODE_RUNNER: Methods and Control Structures - indexOf and substring 
public class Main {
    /** The secret string. */
    private String secret;
    
    /** Constructs a WordMatch object with the given secret string of lowercase letters. */
    public Main(String word) {
        secret = word;
    }
    
    // Part (a) method - needed for part (b) to work
    /** Returns a score for guess, as described in part (a).
     * Precondition: 0 < guess.length() <= secret.length()
     */
    public int scoreGuess(String guess) {
        int count = 0;
        int index = 0;
        
        while (index <= secret.length() - guess.length()) {
            if (secret.substring(index, index + guess.length()).equals(guess)) {
                count++;
            }
            index++;
        }
        
        return count * guess.length() * guess.length();
    }
    
    // ==================== PART (b) SOLUTION ====================
    /** Returns the better of two guesses, as determined by scoreGuess and the rules for a
     * tie-breaker that are described in part (b).
     * Precondition: guess1 and guess2 contain all lowercase letters.
     * guess1 is not the same as guess2.
     */
    public String findBetterGuess(String guess1, String guess2) {
        int score1 = scoreGuess(guess1);
        int score2 = scoreGuess(guess2);
        
        if (score1 > score2) {
            return guess1;
        } else if (score2 > score1) {
            return guess2;
        } else {
            // Scores are equal, return alphabetically greater string
            if (guess1.compareTo(guess2) > 0) {
                return guess1;
            } else {
                return guess2;
            }
        }
    }
    
    // ==================== TESTING PART (b) ONLY ====================
    public static void main(String[] args) {
        System.out.println("========================================");
        System.out.println("PART (b): findBetterGuess Method");
        System.out.println("========================================\n");
        
        // Example: WordMatch game = new WordMatch("concatenation");
        System.out.println("Example: WordMatch game = new WordMatch(\"concatenation\");\n");
        Main game = new Main("concatenation");
        
        System.out.println("----------------------------------------");
        System.out.println("Test Case 1: Different Scores");
        System.out.println("----------------------------------------");
        
        int score1 = game.scoreGuess("ten");
        int score2 = game.scoreGuess("nation");
        
        System.out.println("game.scoreGuess(\"ten\") = " + score1);
        System.out.println("  Calculation: 1 × 3 × 3 = 9");
        System.out.println();
        System.out.println("game.scoreGuess(\"nation\") = " + score2);
        System.out.println("  Calculation: 1 × 6 × 6 = 36");
        System.out.println();
        
        String better1 = game.findBetterGuess("ten", "nation");
        System.out.println("game.findBetterGuess(\"ten\", \"nation\") = \"" + better1 + "\"");
        System.out.println();
        System.out.println("Explanation: Since scoreGuess returns 36 for \"nation\" and 9 for \"ten\",");
        System.out.println("             the guess with the greater score, \"nation\", is returned.");
        System.out.println();
        System.out.println("Expected: \"nation\"");
        System.out.println("Result: " + (better1.equals("nation") ? "✓ PASS" : "✗ FAIL"));
        
        System.out.println("\n----------------------------------------");
        System.out.println("Test Case 2: Same Scores (Tie-Breaker)");
        System.out.println("----------------------------------------");
        
        int score3 = game.scoreGuess("con");
        int score4 = game.scoreGuess("cat");
        
        System.out.println("game.scoreGuess(\"con\") = " + score3);
        System.out.println("  Calculation: 1 × 3 × 3 = 9");
        System.out.println();
        System.out.println("game.scoreGuess(\"cat\") = " + score4);
        System.out.println("  Calculation: 1 × 3 × 3 = 9");
        System.out.println();
        
        String better2 = game.findBetterGuess("con", "cat");
        System.out.println("game.findBetterGuess(\"con\", \"cat\") = \"" + better2 + "\"");
        System.out.println();
        System.out.println("Explanation: Since scoreGuess returns 9 for both \"con\" and \"cat\",");
        System.out.println("             the alphabetically greater guess, \"con\", is returned.");
        System.out.println("             ('con'.compareTo('cat') > 0 because 'o' > 'a')");
        System.out.println();
        System.out.println("Expected: \"con\"");
        System.out.println("Result: " + (better2.equals("con") ? "✓ PASS" : "✗ FAIL"));
        
        System.out.println("\n========================================");
        System.out.println("Part (b) Testing Complete!");
        System.out.println("========================================");
    }
}

Main.main(null);
````
