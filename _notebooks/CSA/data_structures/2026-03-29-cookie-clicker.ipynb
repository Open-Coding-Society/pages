## Now time for a game!

Cookie Clicker is a classic game, using the OCS Game Engine.  Additionally the game writes to the DOM to show stat collections to the page.

### How to Play
1. Click start in the game embedded below.
2. Click the cookie to update count

<div><strong>Clicks:</strong> <span id="clicker-box">--</span></div>

```python
%%js

// GAME_RUNNER: Stack Clicker Mini Lesson embed | hide_edit: true

import { GameControl, Leaderboard } from '/assets/js/GameEnginev1.1/essentials/Imports.js';
import Clicker from '/assets/js/GameEnginev1.1/essentials/Clicker.js'

class CookieClicker {
  constructor(gameEnv) {
    const path = gameEnv.path;
    const width = gameEnv.innerWidth;
    const height = gameEnv.innerHeight;

    const sprite_src_clicker = path + "/hacks/cookie-clicker/assets/baseCookie.png";

    const cookie_clicker = {
       id: 'Cookie Clicker',
       greeting: "Click or collide with me to earn points!",
       src: sprite_src_clicker,
       SCALE_FACTOR: 1,
       ANIMATION_RATE: 50,
       pixels: {height: 512, width: 512},
       INIT_POSITION: { x: 0, y: 0},
       orientation: {rows: 1, columns: 1 },
       down: {row: 0, start: 0, columns: 1, wiggle: 0.10 },
       hitbox: { widthPercentage: 0.15, heightPercentage: 0.15 },
       interact: function(clicks) {
           // Update DOM element
           document.getElementById('clicker-box').textContent = clicks;
           console.log('Clicker clicked, total clicks:', this.clcks);
       },
    };

    this.classes = [{ class: Clicker, data: cookie_clicker}];
  }
}
export const gameLevelClasses = [CookieClicker];
export { GameControl };

```
