## Reverse Polish Notation (RPN) & Postfix Evaluation  

### **Understanding Stacks and Queues**  
- **Stack (LIFO - Last In, First Out)**: Think of stacking cards. The last one placed is the first one removed.  
- **Queue (FIFO - First In, First Out)**: Think of a line at a store. The first one in is the first one out.  

---

### **What is Reverse Polish Notation (RPN)?**  
- **Infix Notation**: Standard mathematical notation where operators are between operands. (e.g., `3 + 5 * 8`)  
- **Postfix Notation (RPN)**: Operators come after the operands. (e.g., `35+8*` instead of `(3+5)*8`)  

**Example Conversions:**  
1. `3 * 5` → `35*`  
2. `(3 + 5) * 8` → `35+8*`  

---

### **Postfix Expression Evaluation**  
**Example:** Solve `8 9 + 10 3 * 8 *`  
#### **Step-by-Step Calculation:**
1. `8 9 +` → `17`
2. `10 3 *` → `30`
3. `30 8 *` → `240`
4. Final result: `17 240` (Not combined yet, needs more context)  

**Try this:** Solve `8 2 ^ 8 8 * +`  
#### **Step-by-Step Calculation:**
1. `8 2 ^` → `64` (Exponentiation: `8^2 = 64`)
2. `8 8 *` → `64`
3. `64 64 +` → `128` (Final result)  

---

### **Why Use Postfix Notation?**  
- **Follows PEMDAS naturally** (Parentheses, Exponents, Multiplication/Division, Addition/Subtraction).  
- **Operators go into a stack**, while **numerals go into a queue**.  
- **Easier to evaluate expressions using stacks**, reducing complexity in parsing.

### Evaluate the RPN
- Make new stack
- For every token in queue
  - If token is number: push into stack
  - If token is operator:
    - Take 2 nums from top of the stack
    - Use the operator: [num1] (operator) [num2]
    - Put result into stack
- When stack only has 1 element, you have your answer!

```python
from collections import deque

# ─────────────────────────────────────────────
#  STACK & QUEUE DEMO
# ─────────────────────────────────────────────

def stack_demo():
    print("=== STACK DEMO (LIFO) ===")
    stack = []
    for item in [10, 20, 30, 40]:
        stack.append(item)
        print(f"  append({item}) → stack = {stack}")
    print()
    while stack:
        val = stack.pop()
        print(f"  pop() = {val} → stack = {stack}")

def queue_demo():
    print("\n=== QUEUE DEMO (FIFO) ===")
    queue = deque()
    for item in ["A", "B", "C", "D"]:
        queue.append(item)
        print(f"  enqueue({item}) → queue = {list(queue)}")
    print()
    while queue:
        val = queue.popleft()
        print(f"  dequeue() = {val} → queue = {list(queue)}")

# ─────────────────────────────────────────────
#  CALCULATOR FUNCTIONS
# ─────────────────────────────────────────────

def add(a, b):       return a + b
def subtract(a, b):  return a - b
def multiply(a, b):  return a * b
def divide(a, b):
    if b == 0: raise ValueError("Cannot divide by zero!")
    return a / b
def power(a, b):     return a ** b
def modulo(a, b):    return a % b
def floor_div(a, b): return a // b

def calculator_demo():
    print("\n=== CALCULATOR FUNCTIONS ===")
    tests = [
        ("add",       add,       10, 5),
        ("subtract",  subtract,  10, 5),
        ("multiply",  multiply,  10, 5),
        ("divide",    divide,    10, 5),
        ("power",     power,      2, 8),
        ("modulo",    modulo,    17, 5),
        ("floor_div", floor_div, 17, 5),
    ]
    for name, fn, a, b in tests:
        print(f"  {name}({a}, {b}) = {fn(a, b)}")

# ─────────────────────────────────────────────
#  INFIX → RPN (Shunting-Yard Algorithm)
# ─────────────────────────────────────────────

PRECEDENCE = {"+": 1, "-": 1, "*": 2, "/": 2, "//": 2, "%": 2, "^": 3}
OPERATORS  = set(PRECEDENCE.keys())

def infix_to_rpn(expression: str) -> list:
    tokens = []
    i = 0
    expr = expression.replace(" ", "")
    while i < len(expr):
        if expr[i:i+2] == "//":
            tokens.append("//")
            i += 2
        elif expr[i].isdigit() or expr[i] == ".":
            j = i
            while j < len(expr) and (expr[j].isdigit() or expr[j] == "."):
                j += 1
            tokens.append(float(expr[i:j]))
            i = j
        else:
            tokens.append(expr[i])
            i += 1

    output_queue = deque()
    op_stack = []

    for token in tokens:
        if isinstance(token, float):
            output_queue.append(token)
        elif token in OPERATORS:
            while (op_stack and op_stack[-1] != "(" and
                   op_stack[-1] in OPERATORS and
                   PRECEDENCE[op_stack[-1]] >= PRECEDENCE[token]):
                output_queue.append(op_stack.pop())
            op_stack.append(token)
        elif token == "(":
            op_stack.append(token)
        elif token == ")":
            while op_stack and op_stack[-1] != "(":
                output_queue.append(op_stack.pop())
            op_stack.pop()

    while op_stack:
        output_queue.append(op_stack.pop())

    return list(output_queue)

# ─────────────────────────────────────────────
#  RPN EVALUATOR
# ─────────────────────────────────────────────

def evaluate_rpn(tokens: list, verbose=True) -> float:
    stack = []
    ops = {
        "+":  lambda a, b: a + b,
        "-":  lambda a, b: a - b,
        "*":  lambda a, b: a * b,
        "/":  lambda a, b: a / b,
        "//": lambda a, b: a // b,
        "%":  lambda a, b: a % b,
        "^":  lambda a, b: a ** b,
    }
    if verbose: print(f"  Tokens: {tokens}")
    for token in tokens:
        if isinstance(token, (int, float)):
            stack.append(token)
            if verbose: print(f"  PUSH {token} → stack = {stack}")
        elif token in ops:
            b = stack.pop()
            a = stack.pop()
            result = ops[token](a, b)
            stack.append(result)
            if verbose: print(f"  {a} {token} {b} = {result} → stack = {stack}")
    if verbose: print(f" Answer: {stack[0]}")
    return stack[0]

# ─────────────────────────────────────────────
#  FULL CALCULATOR (Infix string → Answer)
# ─────────────────────────────────────────────

def rpn_calculator(expression: str) -> float:
    print(f"\n Expression: {expression}")
    rpn = infix_to_rpn(expression)
    print(f"   RPN tokens:  {rpn}")
    result = evaluate_rpn(rpn)
    print(f" Final Answer: {result}")
    return result

# ─────────────────────────────────────────────
#  POSTFIX → INFIX CONVERTER
# ─────────────────────────────────────────────

def postfix_to_infix(tokens: list) -> str:
    stack = []
    for token in tokens:
        if token not in OPERATORS:
            v = int(token) if isinstance(token, float) and token.is_integer() else token
            stack.append(str(v))
        else:
            b = stack.pop()
            a = stack.pop()
            stack.append(f"({a} {token} {b})")
    return stack[0]

def popcorn_hack():
    print("\n=== POPCORN HACK: Postfix → Infix ===")
    problems = [
        ([6, 3, "*", 4, "+"],           "6 3 * 4 +"),
        ([10, 2, 8, "*", "+", 3, "-"],  "10 2 8 * + 3 -"),
        ([15, 3, "/", 4, 2, "*", "+"],  "15 3 / 4 2 * +"),
        ([7, 3, 2, "*", "+", 5, "-"],   "7 3 2 * + 5 -"),
        ([9, 3, "+", 2, "^"],           "9 3 + 2 ^"),
    ]
    simple_ops = {"+": lambda a,b: a+b, "-": lambda a,b: a-b,
                  "*": lambda a,b: a*b, "/": lambda a,b: a/b, "^": lambda a,b: a**b}

    print(f"  {'#':<3} {'Postfix':<28} {'Infix':<32} {'Value'}")
    print("  " + "-"*68)
    for i, (tokens, postfix_str) in enumerate(problems, 1):
        infix = postfix_to_infix(tokens)
        stk = []
        for t in tokens:
            if t in simple_ops: b=stk.pop(); a=stk.pop(); stk.append(simple_ops[t](a,b))
            else: stk.append(t)
        print(f"  {i:<3} {postfix_str:<28} {infix:<32} = {stk[0]}")

# ─────────────────────────────────────────────
#  RUN EVERYTHING
# ─────────────────────────────────────────────

stack_demo()
queue_demo()
calculator_demo()

print("\n=== RPN EVALUATOR EXAMPLES ===")
print("--- 8 9 + 10 3 * 8 * ---")
evaluate_rpn([8, 9, "+", 10, 3, "*", 8, "*"])
print("\n--- 8 2 ^ 8 8 * + ---")
evaluate_rpn([8, 2, "^", 8, 8, "*", "+"])

print("\n=== FULL CALCULATOR ===")
for expr in ["(3 + 5) * 8", "8^2 + 8*8", "10 + 2 * 8 - 3", "(9 + 3)^2"]:
    rpn_calculator(expr)
```
