## Undecidable Problems

---

## What is an Undecidable Problem?
An **undecidable problem** is a problem where no algorithm can always give a correct yes or no answer **for every possible input**.

This does not mean computers are slow or not powerful enough — it means it is **mathematically impossible** to write such a program, no matter how good the programmer is.

Even if a problem is very difficult, it can still be **decidable**.  
**Undecidable means impossible, not just hard.**

---

## The Halting Problem
The most famous undecidable problem is the **halting problem**:

> Can you write a program that looks at any other program and its input, and correctly determines whether that program will eventually stop (halt) or run forever?

The answer is **no**. It has been proven that no such program can exist.

For example, a program with a loop might run forever—or it might stop depending on conditions. There is no general algorithm that can correctly decide this for all programs.

Think of it this way — if you tried to build that program, you could always construct a case that breaks it, causing a contradiction.

---

## Decidable vs Undecidable

| Type | Definition | Example |
|------|------------|---------|
| **Decidable** | An algorithm always gives a correct yes/no answer for every input | Is this number even? |
| **Undecidable** | No algorithm can always give a correct yes/no answer for every input | Will this program halt or run forever? |

---

## Why This Matters
- It shows that **some problems cannot be solved by algorithms**
- It helps us understand the **limits of computing**
- It explains why we **can’t automate every possible task**

```python
%%html
<!-- UI_RUNNER: Paradox Lab interactive undecidable problems game -->
<div id="paradox-lab-game"></div>

<script>
(() => {
  const mount = document.getElementById('paradox-lab-game');
  if (!mount) return;

  mount.innerHTML = `
    <style>
      #paradox-lab-game {
        --bg: #1f1308;
        --panel: rgba(49, 30, 17, 0.92);
        --panel-soft: rgba(77, 48, 27, 0.88);
        --line: rgba(255, 219, 177, 0.16);
        --text: #fff5e8;
        --muted: #d9c3ad;
        --accent: #ff9f1c;
        --accent-2: #2ec4b6;
        --accent-3: #ffbf69;
        --danger: #ff5d73;
        --success: #7bd389;
        --shadow: 0 24px 56px rgba(0, 0, 0, 0.28);
        color: var(--text);
        font-family: "Trebuchet MS", "Avenir Next", sans-serif;
        background:
          radial-gradient(circle at 10% 12%, rgba(255, 159, 28, 0.22), transparent 24%),
          radial-gradient(circle at 90% 0%, rgba(46, 196, 182, 0.18), transparent 26%),
          linear-gradient(145deg, #1a0f07 0%, #26160a 36%, #170d06 100%);
        width: min(100%, 1380px);
        margin: 0 auto;
        border-radius: 30px;
        border: 1px solid rgba(255, 230, 204, 0.08);
        padding: clamp(16px, 2vw, 24px);
        box-shadow: var(--shadow);
        overflow: hidden;
      }
      #paradox-lab-game * { box-sizing: border-box; }
      #paradox-lab-game h3,
      #paradox-lab-game h4,
      #paradox-lab-game p,
      #paradox-lab-game pre {
        margin: 0;
      }
      #paradox-lab-game .hero {
        display: flex;
        justify-content: space-between;
        gap: 16px;
        align-items: flex-start;
        flex-wrap: wrap;
        margin-bottom: 16px;
      }
      #paradox-lab-game .eyebrow {
        text-transform: uppercase;
        letter-spacing: 0.16em;
        font-size: 0.73rem;
        color: var(--accent-2);
        margin-bottom: 8px;
      }
      #paradox-lab-game .hero h3 {
        font-size: clamp(1.9rem, 3vw, 2.7rem);
        line-height: 1.02;
        margin-bottom: 8px;
      }
      #paradox-lab-game .hero p {
        max-width: 64ch;
        color: var(--muted);
        line-height: 1.5;
      }
      #paradox-lab-game .scorebox {
        min-width: min(100%, 320px);
        flex: 1 1 300px;
        background: rgba(255,255,255,0.04);
        border-radius: 20px;
        border: 1px solid rgba(255,255,255,0.08);
        padding: 14px;
      }
      #paradox-lab-game .scoregrid {
        display: grid;
        grid-template-columns: repeat(3, minmax(0, 1fr));
        gap: 10px;
      }
      #paradox-lab-game .scoretile {
        background: rgba(255,255,255,0.04);
        border-radius: 16px;
        padding: 10px;
        border: 1px solid rgba(255,255,255,0.06);
      }
      #paradox-lab-game .scoretile span {
        display: block;
        font-size: 0.76rem;
        color: var(--muted);
        margin-bottom: 4px;
      }
      #paradox-lab-game .scoretile strong {
        font-size: 1.45rem;
      }
      #paradox-lab-game .lab-shell {
        display: grid;
        grid-template-columns: minmax(0, 370px) minmax(0, 1fr);
        gap: 18px;
        align-items: start;
      }
      #paradox-lab-game .panel {
        background: var(--panel);
        border: 1px solid var(--line);
        border-radius: 24px;
        padding: 18px;
        backdrop-filter: blur(12px);
      }
      #paradox-lab-game .section-title {
        font-size: 1rem;
        margin-bottom: 10px;
      }
      #paradox-lab-game .section-copy {
        color: var(--muted);
        line-height: 1.45;
        font-size: 0.92rem;
        margin-bottom: 14px;
      }
      #paradox-lab-game .rule-grid {
        display: grid;
        gap: 10px;
      }
      #paradox-lab-game .rule-card,
      #paradox-lab-game .mission,
      #paradox-lab-game .stat-card,
      #paradox-lab-game .result-card,
      #paradox-lab-game .log,
      #paradox-lab-game .program-card,
      #paradox-lab-game .deck-card {
        background: rgba(255,255,255,0.04);
        border: 1px solid rgba(255,255,255,0.07);
        border-radius: 18px;
      }
      #paradox-lab-game .rule-card {
        padding: 12px;
        transition: transform 140ms ease, border-color 140ms ease, background 140ms ease;
      }
      #paradox-lab-game .rule-card.installed {
        border-color: rgba(123, 211, 137, 0.45);
        background: rgba(123, 211, 137, 0.10);
      }
      #paradox-lab-game .rule-card h4,
      #paradox-lab-game .program-card h4,
      #paradox-lab-game .deck-card h4,
      #paradox-lab-game .result-card h4 {
        font-size: 1rem;
        margin-bottom: 6px;
      }
      #paradox-lab-game .tiny {
        font-size: 0.82rem;
        color: var(--muted);
        line-height: 1.4;
      }
      #paradox-lab-game .pill-row,
      #paradox-lab-game .button-row,
      #paradox-lab-game .badge-row {
        display: flex;
        flex-wrap: wrap;
        gap: 10px;
      }
      #paradox-lab-game .badge,
      #paradox-lab-game .pill {
        display: inline-flex;
        align-items: center;
        justify-content: center;
        border-radius: 999px;
        padding: 0.38rem 0.72rem;
        font-size: 0.8rem;
        border: 1px solid rgba(255,255,255,0.08);
        background: rgba(255,255,255,0.05);
      }
      #paradox-lab-game .pill.hot {
        background: rgba(255, 93, 115, 0.12);
        color: #ffd5dc;
        border-color: rgba(255, 93, 115, 0.3);
      }
      #paradox-lab-game button {
        border: 0;
        border-radius: 999px;
        padding: 0.82rem 1rem;
        font-weight: 800;
        cursor: pointer;
        transition: transform 120ms ease, opacity 120ms ease, box-shadow 120ms ease;
      }
      #paradox-lab-game button:hover { transform: translateY(-1px); }
      #paradox-lab-game button:disabled {
        opacity: 0.45;
        cursor: not-allowed;
        transform: none;
      }
      #paradox-lab-game .primary {
        background: linear-gradient(135deg, #ffbf69, #ff9f1c);
        color: #241307;
        box-shadow: 0 12px 24px rgba(255, 159, 28, 0.28);
      }
      #paradox-lab-game .secondary {
        background: rgba(255,255,255,0.08);
        color: var(--text);
        border: 1px solid rgba(255,255,255,0.08);
      }
      #paradox-lab-game .danger {
        background: linear-gradient(135deg, #ff7b89, #ff5d73);
        color: #28080d;
        box-shadow: 0 12px 24px rgba(255, 93, 115, 0.28);
      }
      #paradox-lab-game .teal {
        background: linear-gradient(135deg, #7de7dc, #2ec4b6);
        color: #08231f;
        box-shadow: 0 12px 24px rgba(46, 196, 182, 0.24);
      }
      #paradox-lab-game .arena {
        display: grid;
        gap: 14px;
      }
      #paradox-lab-game .scanner {
        background: var(--panel-soft);
        border: 1px solid var(--line);
        border-radius: 26px;
        padding: 16px;
        overflow: hidden;
        position: relative;
      }
      #paradox-lab-game .scanner::before {
        content: "";
        position: absolute;
        inset: 0;
        background:
          linear-gradient(90deg, transparent, rgba(255,255,255,0.08), transparent),
          radial-gradient(circle at top right, rgba(46, 196, 182, 0.18), transparent 24%);
        transform: translateX(-100%);
        transition: transform 800ms ease;
        pointer-events: none;
      }
      #paradox-lab-game .scanner.scanning::before {
        transform: translateX(100%);
      }
      #paradox-lab-game .program-card {
        padding: 16px;
        margin-bottom: 12px;
      }
      #paradox-lab-game .program-card.mirror {
        border-color: rgba(255, 93, 115, 0.38);
        background: rgba(255, 93, 115, 0.08);
      }
      #paradox-lab-game .program-head {
        display: flex;
        justify-content: space-between;
        gap: 10px;
        align-items: flex-start;
        margin-bottom: 10px;
      }
      #paradox-lab-game pre {
        white-space: pre-wrap;
        line-height: 1.45;
        font-size: 0.92rem;
        color: #ffe7d0;
        font-family: "SFMono-Regular", Consolas, monospace;
        background: rgba(0,0,0,0.18);
        border-radius: 14px;
        padding: 12px;
        border: 1px solid rgba(255,255,255,0.05);
      }
      #paradox-lab-game .arena-grid {
        display: grid;
        grid-template-columns: minmax(0, 1.15fr) minmax(0, 0.85fr);
        gap: 14px;
      }
      #paradox-lab-game .result-card,
      #paradox-lab-game .deck-card,
      #paradox-lab-game .log {
        padding: 14px;
      }
      #paradox-lab-game .result-banner {
        margin-top: 10px;
        border-radius: 16px;
        padding: 14px;
        font-weight: 700;
        line-height: 1.45;
      }
      #paradox-lab-game .result-banner.halt {
        background: rgba(123, 211, 137, 0.12);
        border: 1px solid rgba(123, 211, 137, 0.28);
        color: #dff9e4;
      }
      #paradox-lab-game .result-banner.loop {
        background: rgba(255, 191, 105, 0.12);
        border: 1px solid rgba(255, 191, 105, 0.28);
        color: #fff0d7;
      }
      #paradox-lab-game .result-banner.unknown {
        background: rgba(255,255,255,0.06);
        border: 1px solid rgba(255,255,255,0.1);
        color: var(--text);
      }
      #paradox-lab-game .result-banner.paradox {
        background: rgba(255, 93, 115, 0.12);
        border: 1px solid rgba(255, 93, 115, 0.32);
        color: #ffe2e7;
      }
      #paradox-lab-game .meter {
        height: 14px;
        border-radius: 999px;
        background: rgba(255,255,255,0.08);
        overflow: hidden;
        margin-top: 8px;
      }
      #paradox-lab-game .meter > span {
        display: block;
        height: 100%;
        width: 0%;
        border-radius: inherit;
        transition: width 320ms ease;
      }
      #paradox-lab-game .meter.coverage > span {
        background: linear-gradient(90deg, #ffbf69, #2ec4b6);
      }
      #paradox-lab-game .meter.stability > span {
        background: linear-gradient(90deg, #7bd389, #ffbf69, #ff5d73);
      }
      #paradox-lab-game .mission-list,
      #paradox-lab-game .log-list,
      #paradox-lab-game .deck-list {
        list-style: none;
        padding: 0;
        margin: 12px 0 0;
      }
      #paradox-lab-game .mission,
      #paradox-lab-game .deck-item,
      #paradox-lab-game .log-list li {
        padding: 11px 12px;
        margin-top: 10px;
      }
      #paradox-lab-game .mission.done {
        background: rgba(123, 211, 137, 0.1);
        border-color: rgba(123, 211, 137, 0.32);
      }
      #paradox-lab-game .deck-item {
        border-top: 1px solid rgba(255,255,255,0.08);
      }
      #paradox-lab-game .deck-item:first-child { border-top: 0; }
      #paradox-lab-game .deck-item strong {
        display: block;
        margin-bottom: 4px;
      }
      #paradox-lab-game .truth {
        color: var(--accent-2);
        font-weight: 700;
      }
      #paradox-lab-game .warning {
        margin-top: 12px;
        padding: 12px;
        border-radius: 16px;
        background: rgba(255, 93, 115, 0.08);
        border: 1px solid rgba(255, 93, 115, 0.24);
        color: #ffe1e6;
      }
      @media (max-width: 1220px) {
        #paradox-lab-game .lab-shell,
        #paradox-lab-game .arena-grid {
          grid-template-columns: 1fr;
        }
      }
      @media (max-width: 860px) {
        #paradox-lab-game .hero {
          flex-direction: column;
        }
        #paradox-lab-game .scoregrid {
          grid-template-columns: 1fr;
        }
        #paradox-lab-game .button-row button {
          flex: 1 1 100%;
        }
      }
      @media (max-width: 640px) {
        #paradox-lab-game {
          border-radius: 20px;
          padding: 14px;
        }
        #paradox-lab-game .panel,
        #paradox-lab-game .scanner,
        #paradox-lab-game .program-card,
        #paradox-lab-game .result-card,
        #paradox-lab-game .deck-card,
        #paradox-lab-game .log,
        #paradox-lab-game .rule-card,
        #paradox-lab-game .mission {
          padding: 12px;
        }
      }
    </style>

    <div class="hero">
      <div>
        <div class="eyebrow">3.2 Undecidable Problems</div>
        <h3>Paradox Lab</h3>
        <p>Build a detector for easy cases, then watch a mirror program break any "perfect" one.</p>
      </div>
      <div class="scorebox">
        <div class="tiny" style="margin-bottom: 10px;">Lab dashboard</div>
        <div class="scoregrid">
          <div class="scoretile"><span>Coverage</span><strong id="coverage-count">0/6</strong></div>
          <div class="scoretile"><span>Installed rules</span><strong id="rule-count">0</strong></div>
          <div class="scoretile"><span>Stability</span><strong id="stability-count">100%</strong></div>
        </div>
      </div>
    </div>

    <div class="lab-shell">
      <div class="panel">
        <h4 class="section-title">Rule Deck</h4>
        <p class="section-copy">These rules catch easy cases. The trap is thinking that means they can catch every case.</p>
        <div class="rule-grid" id="rule-grid"></div>
        <div class="button-row" style="margin-top: 14px;">
          <button class="primary" id="spawn-program">Load Next Program</button>
          <button class="secondary" id="reset-lab">Reset Lab</button>
        </div>
        <div class="warning" id="perfect-warning">Solve a few easy programs first. Then test a "perfect" decider.</div>

        <h4 class="section-title" style="margin-top: 18px;">Missions</h4>
        <div id="mission-list"></div>
      </div>

      <div class="arena">
        <div class="scanner" id="scanner-shell">
          <div class="program-card" id="program-card"></div>
          <div class="button-row" id="action-row"></div>
        </div>

        <div class="arena-grid">
          <div class="result-card">
            <h4>Scanner Output</h4>
            <p class="tiny">Your rules catch some cases. The mirror program breaks the fake perfect one.</p>
            <div class="result-banner unknown" id="result-banner">Load a program to start the lab.</div>
            <div style="margin-top: 14px;">
              <div class="tiny">Coverage meter</div>
              <div class="meter coverage"><span id="coverage-bar"></span></div>
            </div>
            <div style="margin-top: 14px;">
              <div class="tiny">Stability meter</div>
              <div class="meter stability"><span id="stability-bar"></span></div>
            </div>
            <div class="warning" id="lesson-card" style="display: none;"></div>
          </div>

          <div>
            <div class="deck-card">
              <h4>Program Zoo</h4>
              <p class="tiny">Start with these easy programs.</p>
              <div id="deck-list"></div>
            </div>
            <div class="log" style="margin-top: 14px;">
              <h4>Lab Log</h4>
              <ul class="log-list" id="log-list"></ul>
            </div>
          </div>
        </div>
      </div>
    </div>
  `;

  const rules = [
    {
      id: 'explicit-stop',
      name: 'Stop Beacon',
      description: 'Recognizes code with a guaranteed stop command at the end.',
      match: (program) => program.tags.includes('explicit-stop'),
      result: 'HALT'
    },
    {
      id: 'countdown',
      name: 'Countdown Lens',
      description: 'Recognizes loops that shrink to zero and must finish.',
      match: (program) => program.tags.includes('countdown'),
      result: 'HALT'
    },
    {
      id: 'forever-loop',
      name: 'Infinite Siren',
      description: 'Recognizes obvious forever-loops like while(true).',
      match: (program) => program.tags.includes('forever-loop'),
      result: 'LOOP'
    },
    {
      id: 'bounded-repeat',
      name: 'Bounded Loop Meter',
      description: 'Recognizes loops with a fixed repeat count.',
      match: (program) => program.tags.includes('bounded-repeat'),
      result: 'HALT'
    }
  ];

  const ordinaryPrograms = [
    {
      id: 'snack-counter',
      title: 'Snack Counter',
      vibe: 'counts tickets down to zero',
      truth: 'HALT',
      tags: ['countdown'],
      code: `tickets = 4\nwhile (tickets > 0) {\n  tickets = tickets - 1\n}\nprint("closed")`,
      explanation: 'The loop variable moves toward zero, so this one halts.'
    },
    {
      id: 'meteor-alarm',
      title: 'Meteor Alarm',
      vibe: 'contains an obvious forever-loop',
      truth: 'LOOP',
      tags: ['forever-loop'],
      code: `while (true) {\n  flashAlert()\n}`,
      explanation: 'This program never reaches a stopping point.'
    },
    {
      id: 'robot-wave',
      title: 'Robot Wave',
      vibe: 'repeats a fixed number of times',
      truth: 'HALT',
      tags: ['bounded-repeat'],
      code: `repeat 8 times {\n  robot.wave()\n}\nprint("done")`,
      explanation: 'A fixed repeat count makes the runtime finite.'
    },
    {
      id: 'confetti-finale',
      title: 'Confetti Finale',
      vibe: 'ends with a direct stop command',
      truth: 'HALT',
      tags: ['explicit-stop'],
      code: `launchConfetti()\nplaySong()\nstop`,
      explanation: 'The code includes an explicit stop signal.'
    },
    {
      id: 'maze-runner',
      title: 'Maze Runner',
      vibe: 'searches a finite grid',
      truth: 'HALT',
      tags: ['bounded-repeat'],
      code: `for each row in 12x12Grid {\n  for each cell in row {\n    check(cell)\n  }\n}\nprint("searched")`,
      explanation: 'A finite grid means the search must eventually finish.'
    },
    {
      id: 'sparkler-show',
      title: 'Sparkler Show',
      vibe: 'obvious countdown with a celebration at the end',
      truth: 'HALT',
      tags: ['countdown', 'explicit-stop'],
      code: `seconds = 3\nwhile (seconds > 0) {\n  seconds = seconds - 1\n}\nstop`,
      explanation: 'Each step reduces the counter, so the loop cannot continue forever.'
    }
  ];

  const mirrorProgram = {
    id: 'mirror-self',
    title: 'Mirror(Self)',
    vibe: 'reads the perfect decider and does the opposite',
    truth: 'PARADOX',
    tags: ['mirror'],
    code: `Mirror(Self):\n  if PerfectDecider says "HALT" about Mirror(Self):\n    loop forever\n  else:\n    halt immediately`,
    explanation: 'This program is built to contradict any claimed perfect halting detector.'
  };

  const state = {
    installedRules: new Set(),
    seen: new Set(),
    resolved: new Set(),
    currentIndex: 0,
    currentProgram: null,
    perfectUnlocked: false,
    perfectInstalled: false,
    stability: 100,
    log: [
      'Lab ready. Install rules and test easy programs.'
    ],
    lastPrediction: null
  };

  const ui = {
    ruleGrid: mount.querySelector('#rule-grid'),
    coverageCount: mount.querySelector('#coverage-count'),
    ruleCount: mount.querySelector('#rule-count'),
    stabilityCount: mount.querySelector('#stability-count'),
    missionList: mount.querySelector('#mission-list'),
    programCard: mount.querySelector('#program-card'),
    actionRow: mount.querySelector('#action-row'),
    resultBanner: mount.querySelector('#result-banner'),
    coverageBar: mount.querySelector('#coverage-bar'),
    stabilityBar: mount.querySelector('#stability-bar'),
    deckList: mount.querySelector('#deck-list'),
    logList: mount.querySelector('#log-list'),
    lessonCard: mount.querySelector('#lesson-card'),
    scannerShell: mount.querySelector('#scanner-shell'),
    perfectWarning: mount.querySelector('#perfect-warning'),
    spawnProgram: mount.querySelector('#spawn-program'),
    resetLab: mount.querySelector('#reset-lab')
  };

  function addLog(message) {
    state.log = [message, ...state.log].slice(0, 6);
  }

  function installRule(ruleId) {
    if (state.installedRules.has(ruleId)) {
      addLog('That rule is already installed. Try it on another program.');
      render();
      return;
    }
    state.installedRules.add(ruleId);
    addLog(`Installed ${rules.find((rule) => rule.id === ruleId).name}.`);
    render();
  }

  function nextProgram() {
    if (state.currentIndex >= ordinaryPrograms.length) {
      state.currentIndex = 0;
    }
    state.currentProgram = ordinaryPrograms[state.currentIndex];
    state.seen.add(state.currentProgram.id);
    state.currentIndex += 1;
    state.lastPrediction = null;
    addLog(`Loaded ${state.currentProgram.title}.`);
    render();
  }

  function analyzeWithRules(program) {
    for (const rule of rules) {
      if (state.installedRules.has(rule.id) && rule.match(program)) {
        return {
          type: rule.result,
          text: `${rule.name} matched this pattern, so the lab predicts ${rule.result}. ${program.explanation}`
        };
      }
    }
    return {
      type: 'UNKNOWN',
      text: 'No rule matched. Your detector has limits.'
    };
  }

  function runAnalysis() {
    if (!state.currentProgram) {
      addLog('Load a program first.');
      render();
      return;
    }

    ui.scannerShell.classList.add('scanning');
    setTimeout(() => ui.scannerShell.classList.remove('scanning'), 850);

    if (state.currentProgram.id === 'mirror-self' && state.perfectInstalled) {
      addLog('Mirror(Self) is waiting for a prediction lever.');
      render();
      return;
    }

    const result = analyzeWithRules(state.currentProgram);
    state.lastPrediction = result;
    if (result.type !== 'UNKNOWN' && state.currentProgram.truth === result.type) {
      state.resolved.add(state.currentProgram.id);
      addLog(`${state.currentProgram.title} was classified successfully.`);
    } else if (result.type !== 'UNKNOWN') {
      state.stability = Math.max(40, state.stability - 10);
      addLog(`${state.currentProgram.title} exposed a bad rule. Stability dropped.`);
    } else {
      addLog(`${state.currentProgram.title} stayed unresolved. Partial detectors miss cases.`);
    }
    maybeUnlockPerfectMode();
    render();
  }

  function maybeUnlockPerfectMode() {
    if (!state.perfectUnlocked && state.resolved.size >= 4) {
      state.perfectUnlocked = true;
      addLog('A fake perfect decider is now available.');
    }
  }

  function installPerfectDecider() {
    if (!state.perfectUnlocked) return;
    state.perfectInstalled = true;
    state.currentProgram = mirrorProgram;
    state.lastPrediction = {
      type: 'UNKNOWN',
      text: 'PerfectDecider installed. Test it on Mirror(Self).'
    };
    addLog('PerfectDecider installed. Mirror(Self) loaded.');
    render();
  }

  function triggerParadox(prediction) {
    state.stability = 0;
    state.lastPrediction = {
      type: 'PARADOX',
      text: prediction === 'HALT'
        ? 'You told PerfectDecider that Mirror(Self) halts. Mirror(Self) reads that prediction and loops forever instead. The decider is wrong.'
        : 'You told PerfectDecider that Mirror(Self) loops forever. Mirror(Self) reads that prediction and halts immediately instead. The decider is wrong.'
    };
    addLog(`Prediction ${prediction} caused a contradiction. PerfectDecider cannot exist.`);
    ui.lessonCard.style.display = 'block';
    ui.lessonCard.textContent = 'This is the halting problem: special-case detectors can exist, but no perfect decider can work for every program.';
    render();
  }

  function resetLab() {
    state.installedRules = new Set();
    state.seen = new Set();
    state.resolved = new Set();
    state.currentIndex = 0;
    state.currentProgram = null;
    state.perfectUnlocked = false;
    state.perfectInstalled = false;
    state.stability = 100;
    state.log = ['Lab ready. Install rules and test easy programs.'];
    state.lastPrediction = null;
    ui.lessonCard.style.display = 'none';
    ui.lessonCard.textContent = '';
    render();
  }

  function renderRules() {
    ui.ruleGrid.innerHTML = '';
    rules.forEach((rule) => {
      const card = document.createElement('div');
      card.className = `rule-card ${state.installedRules.has(rule.id) ? 'installed' : ''}`;
      card.innerHTML = `
        <h4>${rule.name}</h4>
        <p class="tiny">${rule.description}</p>
        <div class="button-row" style="margin-top: 10px;">
          <button class="${state.installedRules.has(rule.id) ? 'secondary' : 'teal'}" ${state.installedRules.has(rule.id) ? 'disabled' : ''}>${state.installedRules.has(rule.id) ? 'Installed' : 'Install Rule'}</button>
        </div>
      `;
      card.querySelector('button').addEventListener('click', () => installRule(rule.id));
      ui.ruleGrid.appendChild(card);
    });
  }

  function renderProgram() {
    if (!state.currentProgram) {
      ui.programCard.className = 'program-card';
      ui.programCard.innerHTML = `
        <h4>No program loaded</h4>
        <p class="tiny">Click <strong>Load Next Program</strong> to feed the scanner an ordinary program from the lab.</p>
      `;
      return;
    }

    ui.programCard.className = `program-card ${state.currentProgram.id === 'mirror-self' ? 'mirror' : ''}`;
    ui.programCard.innerHTML = `
      <div class="program-head">
        <div>
          <h4>${state.currentProgram.title}</h4>
          <p class="tiny">${state.currentProgram.vibe}</p>
        </div>
        <div class="badge-row">
          <div class="badge ${state.currentProgram.id === 'mirror-self' ? 'pill hot' : 'pill'}">${state.currentProgram.id === 'mirror-self' ? 'Self-reference' : 'Ordinary case'}</div>
        </div>
      </div>
      <pre>${state.currentProgram.code}</pre>
    `;
  }

  function renderActions() {
    ui.actionRow.innerHTML = '';

    if (!state.currentProgram) {
      return;
    }

    if (state.currentProgram.id === 'mirror-self' && state.perfectInstalled) {
      const haltBtn = document.createElement('button');
      haltBtn.className = 'danger';
      haltBtn.textContent = 'Tell PerfectDecider: HALT';
      haltBtn.addEventListener('click', () => triggerParadox('HALT'));

      const loopBtn = document.createElement('button');
      loopBtn.className = 'danger';
      loopBtn.textContent = 'Tell PerfectDecider: LOOP';
      loopBtn.addEventListener('click', () => triggerParadox('LOOP'));

      ui.actionRow.appendChild(haltBtn);
      ui.actionRow.appendChild(loopBtn);
      return;
    }

    const analyzeBtn = document.createElement('button');
    analyzeBtn.className = 'primary';
    analyzeBtn.textContent = 'Analyze With Current Rules';
    analyzeBtn.addEventListener('click', runAnalysis);
    ui.actionRow.appendChild(analyzeBtn);

    if (state.perfectUnlocked && !state.perfectInstalled) {
      const perfectBtn = document.createElement('button');
      perfectBtn.className = 'danger';
      perfectBtn.textContent = 'Install Impossible Perfect Decider';
      perfectBtn.addEventListener('click', installPerfectDecider);
      ui.actionRow.appendChild(perfectBtn);
    }
  }

  function renderResult() {
    const result = state.lastPrediction;
    if (!result) {
      ui.resultBanner.className = 'result-banner unknown';
      ui.resultBanner.textContent = 'Load a program, then analyze it with your current rules.';
      return;
    }
    const klass = result.type === 'HALT'
      ? 'halt'
      : result.type === 'LOOP'
        ? 'loop'
        : result.type === 'PARADOX'
          ? 'paradox'
          : 'unknown';
    ui.resultBanner.className = `result-banner ${klass}`;
    ui.resultBanner.textContent = result.text;
  }

  function renderDeck() {
    ui.deckList.innerHTML = '<div class="deck-list"></div>';
    const list = ui.deckList.firstElementChild;
    ordinaryPrograms.forEach((program) => {
      const item = document.createElement('div');
      const seen = state.seen.has(program.id);
      const solved = state.resolved.has(program.id);
      item.className = 'deck-item';
      item.innerHTML = `
        <strong>${program.title}</strong>
        <div class="tiny">${program.vibe}</div>
        <div class="tiny" style="margin-top: 6px;">${solved ? '<span class="truth">Solved by your rules</span>' : seen ? 'Seen, but not fully resolved yet' : 'Not loaded yet'}</div>
      `;
      list.appendChild(item);
    });
  }

  function renderLog() {
    ui.logList.innerHTML = '';
    state.log.forEach((entry) => {
      const li = document.createElement('li');
      li.textContent = entry;
      ui.logList.appendChild(li);
    });
  }

  function renderMissions() {
    const missions = [
      {
        text: 'Install at least 2 practical rules.',
        done: state.installedRules.size >= 2
      },
      {
        text: 'Resolve at least 4 ordinary programs with your rule deck.',
        done: state.resolved.size >= 4
      },
      {
        text: 'Trigger the Mirror(Self) contradiction.',
        done: state.stability === 0
      }
    ];

    ui.missionList.innerHTML = '';
    missions.forEach((mission) => {
      const card = document.createElement('div');
      card.className = `mission ${mission.done ? 'done' : ''}`;
      card.innerHTML = `<strong>${mission.done ? 'Complete' : 'Pending'}</strong><div class="tiny" style="margin-top: 4px;">${mission.text}</div>`;
      ui.missionList.appendChild(card);
    });
  }

  function renderMeters() {
    const coverage = Math.round((state.resolved.size / ordinaryPrograms.length) * 100);
    ui.coverageBar.style.width = `${coverage}%`;
    ui.stabilityBar.style.width = `${state.stability}%`;
    ui.coverageCount.textContent = `${state.resolved.size}/${ordinaryPrograms.length}`;
    ui.ruleCount.textContent = String(state.installedRules.size);
    ui.stabilityCount.textContent = `${state.stability}%`;
  }

  function renderWarnings() {
    if (state.perfectInstalled) {
      ui.perfectWarning.textContent = 'PerfectDecider is live. The only thing left is the mirror test.';
      return;
    }
    if (state.perfectUnlocked) {
      ui.perfectWarning.textContent = 'Your detector looks strong. Now test a fake perfect decider.';
      return;
    }
    ui.perfectWarning.textContent = 'Solve a few easy programs first. Then test a "perfect" decider.';
  }

  function render() {
    renderRules();
    renderProgram();
    renderActions();
    renderResult();
    renderDeck();
    renderLog();
    renderMissions();
    renderMeters();
    renderWarnings();
  }

  ui.spawnProgram.addEventListener('click', nextProgram);
  ui.resetLab.addEventListener('click', resetLab);

  render();
})();
</script>
```
