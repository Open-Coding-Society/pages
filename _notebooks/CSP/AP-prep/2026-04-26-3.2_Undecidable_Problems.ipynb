## Undecidable Problems

---


## What is an Undecidable Problem?
An **undecidable problem** is a problem where no algorithm can always give a correct yes or no answer for every possible input.

This does not mean computers are slow or not powerful enough — it means it is **mathematically impossible** to write such a program, no matter how good the programmer is.

---

## The Halting Problem
The most famous undecidable problem is the **halting problem**:

> Can you write a program that looks at any other program and its input, and correctly determines whether that program will eventually stop (halt) or run forever?

The answer is **no**. It has been proven that no such program can exist.

Think of it this way — if you tried to build that program, you could always construct a case that breaks it, causing a contradiction.

---

## Decidable vs Undecidable
| Type | Definition | Example |
|------|------------|---------|
| **Decidable** | An algorithm always gives a correct yes/no answer | Is this number even? |
| **Undecidable** | No algorithm can always give a correct yes/no answer | Will this program halt or run forever? |

---

## Why This Matters
- It tells us there are **limits to what computers can do** — not every problem is solvable with code
- Some problems are undecidable even though they are easy to describe
- This is one of the most important ideas in all of computer science

## Mini Game - Decidable or Undecidable?

Run the cell below and answer each prompt with:
- `D` for **Decidable**
- `U` for **Undecidable**

The goal is to practice the big idea: **undecidable** does not mean "hard" or "slow". It means **no algorithm can always solve the problem correctly for every possible input**.

```python
import random

challenges = [
    {
        "prompt": "Given any whole number, determine whether it is even.",
        "answer": "D",
        "explanation": "This is decidable because an algorithm can always check whether a number is divisible by 2."
    },
    {
        "prompt": "Given any program and its input, determine whether the program will eventually halt.",
        "answer": "U",
        "explanation": "This is the halting problem, which is undecidable. No algorithm can always get this right for every possible program and input."
    },
    {
        "prompt": "Given a list of numbers, determine whether the average is greater than 50.",
        "answer": "D",
        "explanation": "This is decidable because a program can compute the average in a finite number of steps."
    },
    {
        "prompt": "Given any program and input, determine whether the program will run forever.",
        "answer": "U",
        "explanation": "This is just the halting problem phrased the other way around, so it is also undecidable."
    },
    {
        "prompt": "Given a maze, determine whether there is a path from start to finish.",
        "answer": "D",
        "explanation": "This is decidable because an algorithm can systematically search the maze."
    },
    {
        "prompt": "Given any program, determine whether it will ever print the word 'HELLO'.",
        "answer": "U",
        "explanation": "This is undecidable in general because it asks you to predict a program's behavior for every possible case."
    }
]

def play_undecidable_game(rounds=5):
    selected = random.sample(challenges, k=min(rounds, len(challenges)))
    score = 0

    print("Decidable or Undecidable?\n")
    print("Type D for decidable or U for undecidable.\n")

    for index, challenge in enumerate(selected, start=1):
        print(f"Round {index}: {challenge['prompt']}")
        guess = input("Your answer (D/U): ").strip().upper()

        while guess not in {"D", "U"}:
            guess = input("Please type D or U: ").strip().upper()

        if guess == challenge["answer"]:
            score += 1
            print("Correct.")
        else:
            correct_label = "Decidable" if challenge["answer"] == "D" else "Undecidable"
            print(f"Not quite. Correct answer: {correct_label}.")

        print(challenge["explanation"])
        print()

    print(f"Final score: {score}/{len(selected)}")

    if score == len(selected):
        print("You identified every case correctly.")
    elif score >= len(selected) - 1:
        print("Strong result. You are distinguishing hard problems from undecidable ones.")
    else:
        print("Review the explanations above and focus on this rule: undecidable means no algorithm can always solve every case.")

# Change the number of rounds if you want a longer or shorter game.
play_undecidable_game(5)
```
