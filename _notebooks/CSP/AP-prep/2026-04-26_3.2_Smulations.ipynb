# What is a Simulation?
A **simulation** is a program that models a real-world process or phenomenon.

Simulations are used when the real thing is:
- **Too dangerous** (crash testing, nuclear reactions)
- **Too expensive** (weather modeling, space missions)
- **Too slow** (evolution, climate change over centuries)

Simulations are always **abstractions** — they leave out details, so results may not perfectly match reality.

---

## Why Simulation Results Differ from Reality
1. **Simplifying assumptions** — not every real-world factor is included
2. **Incomplete data** — wrong inputs lead to wrong outputs
3. **Randomness** — two runs of the same simulation can give different results
4. **Missing variables** — you can only model what you think to include

## Mini Simulation - Coin Flip

Change the number inside `run_simulation()` and re-run the cell.

Try: 10, 100, 1000, 10000

Notice how the results get closer to 50/50 as the number grows.
This is a main idea behind simulations -- more trials, more accurate results.

```python
import random

def flip_coin():
    return random.choice(["Heads", "Tails"])

def run_simulation(num_flips):
    heads = 0
    tails = 0

    for i in range(num_flips):
        result = flip_coin()
        if result == "Heads":
            heads += 1
        else:
            tails += 1

    print(f"\nResults after {num_flips} flips:")
    print(f"Heads: {heads} ({round(heads/num_flips*100, 1)}%)")
    print(f"Tails: {tails} ({round(tails/num_flips*100, 1)}%)")
    print(f"\nExpected: 50% each")
    print(f"The more flips you run, the closer you get to 50/50 -- that's the power of simulations.")

# Change this number and re-run the cell to see how results change
run_simulation(10)
```
