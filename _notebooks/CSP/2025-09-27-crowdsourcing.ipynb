## Crowdsourcing
---
### What is Crowdsourcing?
- Crowdsourcing is when a large group of people (the “crowd”) contribute data, ideas, or work—usually through the internet—to solve a problem or create something.

##### Examples(users = data source):

- ``Wikipedia`` - Anyone can edit → articles improve (or sometimes get messed up)

- ``Yelp`` - People leave reviews → helps others decide where to go

- ``reCAPTCHA`` - When you click “I’m not a robot,” you're helping train AI (like identifying objects)

### **Advantages**

- Fast data collection
- Diverse input (many perspectives)
- Often free or low cost
- Scales easily (millions of users)

### **Disadvantages**

- Data can be wrong or biased
- Trolls / spam / bad actors
- Quality control is difficult
- Privacy concerns

### **EXAMPLE Grade Rating System**

```python
ratings = []

def add_grade(score):
    ratings.append(score)

def get_average():
    if len(ratings) == 0:
        return 0
    return sum(ratings) / len(ratings)

# simulate crowd input
add_grade(91.1)
add_grade(92.5)
add_grade(58.3)

print("Average grade:", get_average())
```

### **Explanation**

- Input: user ratings
- Data storage: list (ratings)
- Processing: average calculation
- Output: displayed result
