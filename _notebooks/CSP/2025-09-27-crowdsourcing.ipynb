## Crowdsourcing

---

### What is Crowdsourcing?
Crowdsourcing is the practice of obtaining input, data, or work from a large number of people (the “crowd”), typically through the internet, to solve a problem or create something.

- It allows **problem-solving at scale** (thousands or millions of contributors)
- It relies on **widespread access to computing and the internet**
- Individuals contribute small pieces that combine into a larger solution

---

### Why It Matters (AP CSP Connection)
Crowdsourcing demonstrates how:
- People collaborate using computing systems
- Large-scale data collection becomes possible
- Solutions can be developed faster and more efficiently
- Computing innovations can have both **intended and unintended consequences**

---

### Examples of Crowdsourcing
- **Wikipedia**  
  → Users edit and update articles collaboratively  
  → Benefit: constantly updated knowledge  
  → Risk: misinformation or vandalism  

- **Yelp / Reviews Platforms**  
  → Users provide opinions and ratings  
  → Benefit: helps decision-making  
  → Risk: biased or fake reviews  

- **reCAPTCHA**  
  → Users identify objects (cars, signs, etc.)  
  → Benefit: trains machine learning systems  
  → Hidden effect: users contribute data without always realizing it  

---

### Citizen Science (Important AP Concept)
Citizen science is scientific research conducted in whole or part by distributed individuals, many of whom are not professional scientists.

- People collect and submit real-world data
- Often uses personal devices (phones, computers)
- Example uses:
  - Tracking wildlife
  - Monitoring weather or environmental changes

👉 This is a **type of crowdsourcing applied to science**

---

### Advantages of Crowdsourcing
- Fast and large-scale data collection
- Diverse perspectives and ideas
- Lower cost than traditional methods
- Enables collaboration across locations

---

### Disadvantages / Risks
- Data may be inaccurate or biased
- Presence of trolls or bad actors
- Difficult to verify quality
- Privacy and ethical concerns

---

### Computing Connection
Crowdsourcing depends on computing systems to:
- Collect and store data
- Process large datasets
- Distribute tasks to users
- Share results globally

Without computing, crowdsourcing at this scale would not be possible.

### **EXAMPLE Grade Rating System**

```python
ratings = []

def add_grade(score):
    ratings.append(score)

def get_average():
    if len(ratings) == 0:
        return 0
    return sum(ratings) / len(ratings)

# simulate crowd input
add_grade(91.1)
add_grade(92.5)
add_grade(58.3)

print("Average grade:", get_average())
```

### **Explanation**

- Input: user ratings
- Data storage: list (ratings)
- Processing: average calculation
- Output: displayed result

let current = 0;
let score = 0;
let answered = false;

function loadQuestion() {
  answered = false;
  document.getElementById("nextBtn").style.display = "none";

  const q = questions[current];
  document.getElementById("question").innerText = q.q;

  const answersDiv = document.getElementById("answers");
  answersDiv.innerHTML = "";

  q.choices.forEach((choice, index) => {
    const btn = document.createElement("button");
    btn.innerText = choice;

    btn.style.display = "block";
    btn.style.margin = "8px auto";
    btn.style.padding = "8px";
    btn.style.width = "90%";
    btn.style.border = "1px solid black";
    btn.style.backgroundColor = "white";
    btn.style.color = "black";
    btn.style.cursor = "pointer";

    btn.onclick = () => checkAnswer(btn, index);
    answersDiv.appendChild(btn);
  });
}

function checkAnswer(button, index) {
  if (answered) return;
  answered = true;

  const correct = questions[current].answer;
  const buttons = document.querySelectorAll("#answers button");

  buttons.forEach((btn, i) => {
    if (i === correct) {
      btn.style.backgroundColor = "#c8f7c5";
    } else if (i === index) {
      btn.style.backgroundColor = "#f7c5c5";
    }
    btn.disabled = true;
  });

  if (index === correct) {
    score++;
  }

  document.getElementById("score").innerText =
    "Score: " + score + "/" + questions.length;

  document.getElementById("nextBtn").style.display = "block";
}

function nextQuestion() {
  current++;
  if (current < questions.length) {
    loadQuestion();
  } else {
    document.getElementById("question").innerText = "Finished";
    document.getElementById("answers").innerHTML = "";

    document.getElementById("score").innerText =
      "Final Score: " + score + "/" + questions.length;

    document.getElementById("nextBtn").style.display = "none";
  }
}
