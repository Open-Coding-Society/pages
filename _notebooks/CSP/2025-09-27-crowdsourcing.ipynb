## Crowdsourcing: Solving Problems at Scale

---

## **ENDURING UNDERSTANDING (IOC-1)**

While computing innovations are typically designed to achieve a specific purpose, they may have **unintended consequences**.

---

## **LEARNING OBJECTIVE (IOC-1.E)**

**Explain how people participate in problem-solving processes at scale.**

---

## **ESSENTIAL KNOWLEDGE**

### **IOC-1.E.1: Widespread Access to Information and Public Data**
Widespread access to information and public data facilitates:
- **Identification of problems** - More people can see and understand issues
- **Development of solutions** - Diverse perspectives create better answers
- **Dissemination of results** - Solutions can be shared globally instantly

---

### **IOC-1.E.2 & IOC-1.E.3: Citizen Science**

**Science has been affected by using distributed and "citizen science" to solve scientific problems.**

**Citizen Science Definition:** Scientific research conducted in whole or part by distributed individuals, many of whom may not be professional scientists, who contribute relevant data to research using their own computing devices.

**Examples of Citizen Science:**
- Tracking wildlife migrations via mobile apps
- Monitoring environmental changes (air quality, water pollution)
- Classifying astronomical objects (Galaxy Zoo)
- Reporting disease outbreaks in real-time
- Measuring seismic activity during earthquakes

---

### **IOC-1.E.4: What is Crowdsourcing?**

**Crowdsourcing** is the practice of obtaining input or information from a large number of people via the Internet.

**Key characteristics:**
- Distributed participants (can be thousands or millions)
- Relies on internet connectivity for coordination
- Participants contribute small pieces that combine into solutions
- Can be organized or spontaneous

**Examples of Crowdsourcing:**
- **Wikipedia** → Collaborative knowledge building
- **Google reCAPTCHA** → Data collection for AI training
- **Review platforms (Yelp, Amazon)** → Aggregated opinions
- **Folding@Home** → Computing power for scientific research
- **OpenStreetMap** → Community-built maps

---

### **IOC-1.E.5: Human Capabilities Enhanced by Collaboration via Computing**

Computing tools enable collaboration that would be impossible otherwise:
- **Scale:** Connecting millions of people globally
- **Speed:** Instant sharing and aggregation of data
- **Accessibility:** Participation from anywhere with internet
- **Diversity:** Access to perspectives from different regions and backgrounds
- **Efficiency:** Automated collection, processing, and analysis

---

### **IOC-1.E.6: New Models for Collaboration**

Crowdsourcing offers new business and social models:
- **Funding models:** Crowdfunding connects causes with supporters globally
- **Business innovation:** Companies tap into collective intelligence
- **Social impact:** Nonprofits and activists mobilize support quickly
- **Hybrid expertise:** Professional organizations work with citizen contributors

---

## **Advantages of Crowdsourcing**
- Fast and large-scale data collection
- Diverse perspectives and ideas
- Lower cost than traditional methods
- Enables collaboration across locations and time zones
- Can solve problems too large for single organizations

---

## **Risks and Unintended Consequences**
- Data accuracy and quality concerns
- Presence of bad-faith participants (trolls, manipulators)
- Difficulty verifying contributor credentials
- Privacy and ethical concerns about data use
- **Unintended consequences:** Data misuse, bias amplification, manipulation of public opinion
- Spread of misinformation at scale

---

## **Why Computing is Essential**
Crowdsourcing **depends entirely on computing systems** to:
- **Collect** data from distributed sources
- **Store** massive datasets
- **Process** and analyze information
- **Coordinate** tasks across thousands of people
- **Distribute** results globally
- **Automate** quality checks and aggregation

**Without computing infrastructure, crowdsourcing at this scale would be impossible.**

### **EXAMPLE Grade Rating System**

```python
ratings = []

def add_grade(score):
    ratings.append(score)

def get_average():
    if len(ratings) == 0:
        return 0
    return sum(ratings) / len(ratings)

# simulate crowd input
add_grade(91.1)
add_grade(92.5)
add_grade(58.3)

print("Average grade:", get_average())
```

### **Explanation**

- Input: user ratings
- Data storage: list (ratings)
- Processing: average calculation
- Output: displayed result

let current = 0;
let score = 0;
let answered = false;

function loadQuestion() {
  answered = false;
  document.getElementById("nextBtn").style.display = "none";

  const q = questions[current];
  document.getElementById("question").innerText = q.q;

  const answersDiv = document.getElementById("answers");
  answersDiv.innerHTML = "";

  q.choices.forEach((choice, index) => {
    const btn = document.createElement("button");
    btn.innerText = choice;

    btn.style.display = "block";
    btn.style.margin = "8px auto";
    btn.style.padding = "8px";
    btn.style.width = "90%";
    btn.style.border = "1px solid black";
    btn.style.backgroundColor = "white";
    btn.style.color = "black";
    btn.style.cursor = "pointer";

    btn.onclick = () => checkAnswer(btn, index);
    answersDiv.appendChild(btn);
  });
}

function checkAnswer(button, index) {
  if (answered) return;
  answered = true;

  const correct = questions[current].answer;
  const buttons = document.querySelectorAll("#answers button");

  buttons.forEach((btn, i) => {
    if (i === correct) {
      btn.style.backgroundColor = "#c8f7c5";
    } else if (i === index) {
      btn.style.backgroundColor = "#f7c5c5";
    }
    btn.disabled = true;
  });

  if (index === correct) {
    score++;
  }

  document.getElementById("score").innerText =
    "Score: " + score + "/" + questions.length;

  document.getElementById("nextBtn").style.display = "block";
}

function nextQuestion() {
  current++;
  if (current < questions.length) {
    loadQuestion();
  } else {
    document.getElementById("question").innerText = "Finished";
    document.getElementById("answers").innerHTML = "";

    document.getElementById("score").innerText =
      "Final Score: " + score + "/" + questions.length;

    document.getElementById("nextBtn").style.display = "none";
  }
}
