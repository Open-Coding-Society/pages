## What is a Simulation?

A **simulation** is a program that **models a real-world process** to make predictions or explore outcomes — without doing the thing in real life.

Simulations are used when the real thing is:
- ☠️ **Too dangerous** (crash-testing cars, nuclear reactions)
- 💸 **Too expensive** (space missions, drug trials)
- ⏳ **Too slow** (climate change, evolution)
- 🔁 **Needs repetition** (rolling dice 1 million times)

---

## Key Vocabulary

| Term | Definition |
|------|------------|
| **Simulation** | A model of a real-world process run on a computer |
| **Random number generation** | Producing unpredictable values to mimic chance |
| **Model** | A simplified representation of a system |
| **Bias** | When a simulation doesn't accurately reflect reality |

---

##  Limitations of Simulations

Simulations are **not perfect**. They are only as good as the assumptions built into them:
- They may **oversimplify** reality
- They depend on **accurate data**
- Results carry **uncertainty** — especially with randomness
- A flawed model produces **flawed predictions**

## 🪙 Example 1: Coin Flip Simulation

Let's simulate flipping a coin. In real life, a fair coin should land heads ~50% of the time.  
**Try changing `num_flips` to see how results vary with more or fewer flips!**

```python
import random

# ✏️ CHANGE THIS: try 10, 100, 1000, 10000
num_flips = 100

heads = 0
tails = 0

for _ in range(num_flips):
    flip = random.choice(["Heads", "Tails"])
    if flip == "Heads":
        heads += 1
    else:
        tails += 1

print(f"Flips: {num_flips}")
print(f"Heads: {heads}  ({heads/num_flips*100:.1f}%)")
print(f"Tails: {tails}  ({tails/num_flips*100:.1f}%)")
print()
print("💡 The more flips you do, the closer you get to 50/50!")
print("   This is called the Law of Large Numbers.")
```

## 🎲 Example 2: Dice Roll Simulation

Roll two 6-sided dice many times and see what totals come up most often.  
**Can you predict which total will be most common before running it?**

```python
import random

# ✏️ CHANGE THIS: number of dice rolls to simulate
num_rolls = 1000

totals = {}

for _ in range(num_rolls):
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    total = die1 + die2
    totals[total] = totals.get(total, 0) + 1

print(f"Results after {num_rolls} rolls:\n")
print(f"{'Total':>6} | {'Count':>6} | Bar")
print("-" * 40)
for t in sorted(totals):
    bar = "█" * (totals[t] // (num_rolls // 100))
    print(f"{t:>6} | {totals[t]:>6} | {bar}")

most_common = max(totals, key=totals.get)
print(f"\n🏆 Most common total: {most_common}")
print("💡 Why is 7 usually the winner? It has the most combinations!")
```

## 🌧️ Example 4: Weather Simulation (Biased Random)

Real simulations use **weighted probabilities** to model reality more accurately.  
Here we simulate a week of weather where some outcomes are more likely than others.

**Try editing the probabilities and see what happens to the weekly forecast!**

```python
import random

# ✏️ CHANGE THESE: probabilities must add up to 1.0
weather_options = ["☀️ Sunny", "⛅ Cloudy", "🌧️ Rainy", "⛈️ Stormy"]
probabilities   = [0.50,       0.25,        0.15,        0.10      ]

# ✏️ CHANGE THIS: how many days to simulate?
num_days = 7

print(f"📅 {num_days}-Day Weather Forecast (Simulated)\n")
counts = {w: 0 for w in weather_options}

for day in range(1, num_days + 1):
    weather = random.choices(weather_options, weights=probabilities, k=1)[0]
    counts[weather] += 1
    print(f"  Day {day}: {weather}")

print(f"\n📊 Summary:")
for w, c in counts.items():
    print(f"  {w}: {c} day(s)")

print()
print("💡 Notice: Sunny still dominates because it has 50% probability.")
print("   Real weather models use MUCH more complex probability rules.")
```

## ✅ Key Takeaways

1. **Simulations model real-world processes** when testing in reality is unsafe, costly, or impractical.
2. **Random number generation** is the engine behind most simulations.
3. **More trials = more accurate results** (Law of Large Numbers).
4. Simulations have **limitations** — they simplify reality and rely on the quality of their model.
5. **Bias** in a model leads to unreliable simulation results.
