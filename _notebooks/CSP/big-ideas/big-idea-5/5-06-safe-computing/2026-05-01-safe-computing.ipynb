# Safe Computing - Big Idea 5.6

## Goal

By the end of this lesson, you should be able to explain how computing systems can create privacy and identity risks, then choose defenses that reduce those risks.

## Key Idea

Computing innovations often collect, store, and share data. That data can be useful, but it can also expose people to privacy loss, identity theft, scams, and unsafe access to accounts.


## Terms to Know

| Term | Meaning | Quick Example |
|---|---|---|
| **PII** | Personally identifiable information that can identify, link to, or describe a person | Phone number, medical data, biometric data |
| **Digital footprint** | Data trail created by online actions | Search history, posts, location data |
| **Phishing** | A fake message or site that tricks users into giving private information | Fake password reset email |
| **Malware** | Software designed to damage, spy, steal, or take control | A program that steals files |
| **Keylogger** | Malware or hardware that records keystrokes | Passwords captured while typing |
| **Rogue access point** | Unauthorized Wi-Fi access point used to capture traffic or trick users | Fake school Wi-Fi network |
| **Encryption** | Encoding data so only authorized users can read it | HTTPS protecting web traffic |
| **Authentication** | Verifying that a user is who they claim to be | Password plus phone code |


## Match the Risk to a Defense

| Risk | Defense |
|---|---|
| Someone guesses or steals a password | Use strong passwords and multifactor authentication |
| A website collects too much personal data | Limit data sharing and check privacy settings |
| A message asks for login details | Verify the sender and avoid suspicious links |
| A device has old security bugs | Install software updates |
| Someone intercepts network traffic | Use encrypted connections like HTTPS |
| A public Wi-Fi network looks suspicious | Avoid logging in or use a trusted network |

AP exam tip: security usually reduces risk, but it does not make a system perfectly safe.


## Class Activity: Security Shield Relay

This is a fast physical activity for tables or small groups.

### Materials

- 6 risk cards
- 6 defense cards
- 3 attack cards
- 1 sheet of paper per group

### Setup

Write these cards on paper and cut them out.

**Risk cards:** stolen password, fake Wi-Fi, leaked location, old software, fake login page, overshared profile

**Defense cards:** multifactor authentication, HTTPS, privacy settings, software update, verify sender, avoid unknown networks

**Attack cards:** phishing, malware, rogue access point

### How to Play

1. Each group gets the mixed card pile.
2. They have 2 minutes to match each risk with the best defense.
3. They pick one attack card and write a 2 sentence scenario using it.
4. Groups rotate papers and try to identify the attack and the missing defense.
5. The class discusses which defenses reduce risk and which risks still remain.

### Why It Works

The game makes students practice the AP skill of connecting a computing risk to a real security choice, instead of only memorizing definitions.


## AP Style Check

### Question 1

A fitness app stores a user's name, heart rate history, and location from runs. Which item is the best example of PII?

- A. The app's source code
- B. The user's heart rate history linked to their name
- C. The app's download size
- D. The color of the app icon

**Answer:** B

**Why:** Medical or health-related data connected to a person can identify or describe them.

### Question 2

A student receives an email that looks like it is from the school. It asks them to click a link and re-enter their password. What type of attack is this most likely to be?

- A. Phishing
- B. Compression
- C. Citizen science
- D. Lossless encoding

**Answer:** A

**Why:** Phishing tries to trick users into revealing private information or account credentials.

### Question 3

Which choice best explains why multifactor authentication is safer than only using a password?

- A. It deletes all personal data from the account
- B. It requires another form of proof before access is granted
- C. It prevents every possible attack
- D. It makes encryption unnecessary

**Answer:** B

**Why:** MFA adds another verification step, so a stolen password alone may not be enough to access the account.
