# Safe Computing - Big Idea 5.6


Safe computing is about protecting people, accounts, and data from privacy leaks, scams, and unauthorized access.

The AP exam usually asks you to connect a **risk** to a **defense**.


## Four Things to Know

| Concept | What it means | Fast example |
|---|---|---|
| **PII** | Data that can identify or describe a person | Name, phone number, location, medical data |
| **Phishing** | A fake message that tricks people into giving private info | Fake password reset email |
| **Malware** | Software meant to damage, spy, or steal | Keylogger stealing typed passwords |
| **Encryption** | Scrambling data so outsiders cannot read it | HTTPS on a login page |

**Authentication** checks who you are. **Multifactor authentication** is stronger because it requires more than just a password.


## Risk to Defense

| Risk | Defense |
|---|---|
| Stolen password | Multifactor authentication |
| Fake login email | Verify the sender and avoid suspicious links |
| App collects too much data | Check privacy settings and share less |
| Old security bug | Install software updates |
| Public Wi-Fi interception | Use HTTPS or a trusted network |

**AP tip:** a defense reduces risk. It usually does not remove all risk.


## 60 Second Mini Activity: Scam Detective

Read the three messages and vote for the riskiest one.

**A.** Your school posts: "The Wi-Fi password changed. Ask your teacher for the new one."

**B.** A text says: "Your account will close today. Click this link and enter your password now."

**C.** A website login page shows `https://` and asks for your normal username.

<details>
<summary>Reveal answer</summary>

**B is the riskiest.** It uses urgency, a link, and a password request. That is a phishing warning sign.

Best defense: do not click the link. Go to the real site directly or verify the sender.

</details>


## Quick Check

### 1. PII

A fitness app stores a user's name, heart rate history, and location from runs. Which item is the best example of PII?

- A. The app's source code
- B. The user's heart rate history linked to their name
- C. The app's download size
- D. The color of the app icon
<details>
<summary>Reveal answer</summary>

**Answer:** B

**Why:** Medical or health-related data connected to a person can identify or describe them.

</details>

### 2. Attack Type

A student receives an email that looks like it is from the school. It asks them to click a link and re-enter their password. What type of attack is this most likely to be?

- A. Phishing
- B. Compression
- C. Citizen science
- D. Lossless encoding

<details>
<summary>Reveal answer</summary>

**Answer:** A

**Why:** Phishing tries to trick users into revealing private information or account credentials.

</details>

### 3. Best Defense

Which choice best explains why multifactor authentication is safer than only using a password?

- A. It deletes all personal data from the account
- B. It requires another form of proof before access is granted
- C. It prevents every possible attack
- D. It makes encryption unnecessary

<details>
<summary>Reveal answer</summary>

**Answer:** B

**Why:** MFA adds another verification step, so a stolen password alone may not be enough to access the account.

</details>
