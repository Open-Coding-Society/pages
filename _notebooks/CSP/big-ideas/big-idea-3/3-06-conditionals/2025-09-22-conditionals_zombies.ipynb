## Conditionals Lesson
---
### 1. What is a Conditional?
- A conditional lets a program make decisions.
- Example in plain English:
**“If it’s raining, bring an umbrella. Otherwise, don't bring one.”**

### 2. Basic `if` Statement:
#### Python:


```python
x = 10

if x > 5:
    print("x is greater than 5")
```

#### JavaScript:

```python
%%js

let x = 10;

if (x > 5) {
    console.log("x is greater than 5");
}
```

### 3. `If ... else`
#### Python:

```python
age = 16

if age >= 18:
    print("You can vote")
else:
    print("You are too young to vote")
```

#### JavaScript:

```python
%%js

let age = 16;

if (age >= 18) {
    console.log("You can vote");
} else {
    console.log("You are too young to vote");
}
```

### 4. `If ... else if ... else` (multiple conditions)
#### Common mistakes:
- Mistaking ``elif`` for ``else if``
    - ``elif`` is only used in Python
    - ``else if`` is only used in Javascript
#### Python:

```python
score = 85 

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
else:
    print("Keep working hard!")
```

#### JavaScript:

```python
%%js

let score = 85;

if (score >= 90) {
    console.log("Grade: A");
} else if (score >= 80) {
    console.log("Grade: B");
} else {
    console.log("Keep working hard!");
}
```

### 5. Comparison & Logical Operators
- Comparison:
    - `==` equal
    - `!=` not equal
    - `<`, `>`, `<=`, `>=`
- Logical:
    - Python → `and`, `or`, `not`
    - JavaScript → `&&`, `||`, `!`
#### Python:

```python
temperature = 72

if temperature > 60 and temperature < 80:
    print("Perfect weather!")
```

#### JavaScript:

```python
%%js

let temperature = 72;

if (temperature > 60 && temperature < 80) {
    console.log("Perfect weather!");
}
```

### 6. Booleans in Conditionals
#### Python:

```python
is_raining = True
if is_raining:
    print("Take an umbrella")
else:
    print("No umbrella needed")
```

#### Javascript

```python
%%js

let isRaining = true;
if (isRaining) {
    console.log("Take an umbrella");
} else {
    console.log("No umbrella needed");
}
```

function switchTab(tabName, button) {
    const items = document.querySelectorAll('.content-item');
    items.forEach(item => item.style.display = 'none');

    const buttons = document.querySelectorAll('.tab-button');
    buttons.forEach(btn => btn.classList.remove('active'));

    document.getElementById(tabName).style.display = 'block';
    if (button) button.classList.add('active');

    const headingMap = {
        'activity1': 'Activity 1',
        'activity2': 'Activity 2',
        'activity3': 'Activity 3',
        'activity4': 'Activity 4'
    };
    document.getElementById('header').textContent = headingMap[tabName];
}

let coinCount = 0;
let factIndex = 0;
const gridSize = 5;
const boardState = [];
let playerPosition = { row: 4, col: 0 };
let coinPositions = [];

function randomCoinPositions(count) {
    const positions = [];
    while (positions.length < count) {
        const row = Math.floor(Math.random() * gridSize);
        const col = Math.floor(Math.random() * gridSize);
        const occupied = positions.some(pos => pos.row === row && pos.col === col) || (row === playerPosition.row && col === playerPosition.col);
        if (!occupied) {
            positions.push({ row, col });
        }
    }
    return positions;
}

function initializeBoard() {
    if (coinPositions.length === 0) {
        coinPositions = randomCoinPositions(5);
    }
    const board = document.getElementById('board');
    boardState.length = 0;
    board.innerHTML = '';

    for (let row = 0; row < gridSize; row += 1) {
        for (let col = 0; col < gridSize; col += 1) {
            const cell = document.createElement('div');
            cell.className = 'board-cell';
            cell.dataset.row = row;
            cell.dataset.col = col;
            cell.style.background = '#f8fafc';
            cell.style.borderRadius = '14px';
            cell.style.display = 'flex';
            cell.style.alignItems = 'center';
            cell.style.justifyContent = 'center';
            cell.style.fontSize = '24px';
            cell.style.color = '#0f172a';
            cell.style.transition = 'background 0.2s ease';
            board.appendChild(cell);
            boardState.push({ row, col, element: cell });
        }
    }
    updateBoard();
}

function updateBoard() {
    boardState.forEach(cell => {
        const isPlayer = cell.row === playerPosition.row && cell.col === playerPosition.col;
        const isCoin = coinPositions.some(pos => pos.row === cell.row && pos.col === cell.col);
        cell.element.textContent = isPlayer ? '😀' : isCoin ? '🪙' : '';
        cell.element.style.background = isPlayer ? '#dbeafe' : isCoin ? '#fef3c7' : '#f8fafc';
    });
}

function movePlayer(direction) {
    let { row, col } = playerPosition;
    if (direction === 'left') col = Math.max(0, col - 1);
    if (direction === 'right') col = Math.min(gridSize - 1, col + 1);
    if (direction === 'up') row = Math.max(0, row - 1);
    if (direction === 'down') row = Math.min(gridSize - 1, row + 1);

    const moved = row !== playerPosition.row || col !== playerPosition.col;
    playerPosition = { row, col };
    if (moved) {
        handleCoinPickup();
        updateBoard();
    }
}

function handleCoinPickup() {
    const coinIndex = coinPositions.findIndex(pos => pos.row === playerPosition.row && pos.col === playerPosition.col);
    if (coinIndex !== -1) {
        coinPositions.splice(coinIndex, 1);
        coinCount += 1;
        updateScore();
    }
}

function updateScore() {
    const coinElem = document.getElementById('coin-count');
    const nextFactElem = document.getElementById('next-fact');
    coinElem.textContent = coinCount;

    if (coinCount > 0 && coinCount % 5 === 0 && factIndex < conditionalFacts.length) {
        showPopup(conditionalFacts[factIndex]);
        factIndex += 1;
        updateFactBank();
        if (factIndex < conditionalFacts.length) {
            coinPositions = [];
            initializeBoard();
        }
    }

    if (factIndex >= conditionalFacts.length) {
        nextFactElem.textContent = 'All facts unlocked!';
    } else {
        nextFactElem.textContent = 'Next fact at ' + ((factIndex + 1) * 5) + ' coins';
    }
    updateFactBank();
}

function showPopup(message) {
    const popup = document.getElementById('popup');
    const popupText = document.getElementById('popup-text');
    popupText.textContent = message;
    popup.style.display = 'flex';
}

function closePopup() {
    const popup = document.getElementById('popup');
    popup.style.display = 'none';
}

function toggleFactBank() {
    const list = document.getElementById('fact-bank-list');
    list.style.display = list.style.display === 'none' ? 'block' : 'none';
}

function updateFactBank() {
    const countElem = document.getElementById('fact-count');
    const listElem = document.getElementById('fact-bank-list');
    countElem.textContent = factIndex;
    if (factIndex === 0) {
        listElem.innerHTML = '<div style="font-size: 14px; color: #6b7280;">No facts unlocked yet.</div>';
    } else {
        let html = '';
        for (let i = 0; i < factIndex; i++) {
            html += `<div style="font-size: 14px; color: #374151; margin-bottom: 8px; padding: 8px; background: #f8fafc; border-radius: 8px;">${conditionalFacts[i]}</div>`;
        }
        listElem.innerHTML = html;
    }
}

function setupControls() {
    document.addEventListener('keydown', function(event) {
        const key = event.key.toLowerCase();
        if (key === 'arrowleft' || key === 'a') {
            event.preventDefault();
            movePlayer('left');
        }
        if (key === 'arrowright' || key === 'd') {
            event.preventDefault();
            movePlayer('right');
        }
        if (key === 'arrowup' || key === 'w') {
            event.preventDefault();
            movePlayer('up');
        }
        if (key === 'arrowdown' || key === 's') {
            event.preventDefault();
            movePlayer('down');
        }
    });
}

initializeBoard();
setupControls();
updateScore();
updateFactBank();
