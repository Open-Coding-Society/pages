## Dev Quest Runner

A gamified series of coding challenges built on the GameEngine framework. Each quest teaches a core developer skill — file structure, Makefiles, merge conflicts, and debugging. Complete quests in order to unlock the next one, earn points, and level up.

**How it works:**
- Each quest is a standalone game level with its own objectives
- Players interact with the game world to complete tasks
- Points and XP are awarded on completion
- Quests must be completed in order

````python
%%js

// GAME_RUNNER: Quest 1 - File Structure 101. Walk into each building to learn good practices, answer a quiz, and rack up points!

import { GameControl, GameEnvBackground, Player, NPC } from '/assets/js/GameEnginev1.1/essentials/Imports.js';

class QuestFileStructure {
  static gameScore = { points: 0, buildingsVisited: 0, quizzesCompleted: 0 };

  // ─── Toast Notification ──────────────────────────────────────────────────
  static showToast(msg, color = '#10b981') {
    const toast = document.createElement('div');
    toast.style.cssText = `
      position: fixed; top: 20px; right: 20px; z-index: 99999;
      background: #1a1a2e; border: 2px solid ${color};
      color: ${color}; font-family: 'Courier New', monospace; font-size: 14px;
      padding: 12px 20px; border-radius: 6px; font-weight: bold;
      box-shadow: 0 0 20px ${color}33;
    `;
    toast.textContent = msg;
    document.body.appendChild(toast);
    setTimeout(() => toast.remove(), 2500);
  }

  static showLessonAndQuiz(buildingName, lessonContent, quizQuestion) {
    return new Promise((resolve) => {
      const overlay = document.createElement('div');
      overlay.style.cssText = `
        position: fixed; top: 0; left: 0; width: 100%; height: 100%;
        background: rgba(0,0,0,0.85); z-index: 9999;
        display: flex; align-items: center; justify-content: center;
        font-family: 'Courier New', monospace;
      `;

      const card = document.createElement('div');
      card.style.cssText = `
        background: #1a1a2e; border: 3px solid #00d4ff;
        border-radius: 12px; padding: 32px; max-width: 700px; width: 90%;
        box-shadow: 0 0 40px rgba(0,212,255,0.3); color: #e2e8f0;
      `;

      const title = document.createElement('h2');
      title.style.cssText = 'color: #00d4ff; margin: 0 0 16px 0; font-size: 24px; text-transform: uppercase; letter-spacing: 2px;';
      title.textContent = `📚 ${buildingName}`;

      const content = document.createElement('div');
      content.style.cssText = 'color: #cdd6f4; font-size: 15px; line-height: 1.8; margin-bottom: 20px;';
      content.innerHTML = lessonContent;

      const continueBtn = document.createElement('button');
      continueBtn.style.cssText = `
        background: #00d4ff; color: #000; border: none; padding: 12px 28px;
        border-radius: 6px; cursor: pointer; font-family: inherit; font-size: 14px;
        font-weight: bold; width: 100%; margin-top: 16px;
      `;
      continueBtn.textContent = 'Continue to Quiz →';

      continueBtn.onclick = () => {
        overlay.remove();
        QuestFileStructure.showQuiz(buildingName, quizQuestion, resolve);
      };

      card.appendChild(title);
      card.appendChild(content);
      card.appendChild(continueBtn);
      overlay.appendChild(card);
      document.body.appendChild(overlay);
    });
  }

  // ─── Multiple Choice Quiz ───────────────────────────────────────────────
  static showQuiz(buildingName, quizData, onComplete) {
    const overlay = document.createElement('div');
    overlay.style.cssText = `
      position: fixed; top: 0; left: 0; width: 100%; height: 100%;
      background: rgba(0,0,0,0.85); z-index: 9999;
      display: flex; align-items: center; justify-content: center;
    `;

    const card = document.createElement('div');
    card.style.cssText = `
      background: #1a1a2e; border: 3px solid #f38ba8;
      border-radius: 12px; padding: 32px; max-width: 600px; width: 90%;
      box-shadow: 0 0 40px rgba(243,139,168,0.3); color: #e2e8f0;
      font-family: 'Courier New', monospace;
    `;

    const title = document.createElement('h3');
    title.style.cssText = 'color: #f38ba8; margin: 0 0 12px 0; font-size: 18px; text-transform: uppercase; letter-spacing: 1px;';
    title.textContent = `⚡ Quiz: ${buildingName}`;

    const question = document.createElement('p');
    question.style.cssText = 'color: #e2e8f0; font-size: 15px; margin: 0 0 18px 0; line-height: 1.6;';
    question.textContent = quizData.question;

    const buttonWrapper = document.createElement('div');
    buttonWrapper.style.cssText = 'display: flex; flex-direction: column; gap: 10px;';

    const answers = [...quizData.options].sort(() => Math.random() - 0.5);

    answers.forEach((ans) => {
      const btn = document.createElement('button');
      btn.style.cssText = `
        background: #2a2a3e; border: 2px solid #444;
        color: #cdd6f4; padding: 12px 16px; text-align: left; 
        cursor: pointer; border-radius: 6px; font-size: 14px;
        font-family: inherit; transition: all 0.2s;
      `;
      btn.textContent = ans.text;

      btn.onmouseenter = () => {
        btn.style.borderColor = '#00d4ff';
        btn.style.boxShadow = '0 0 12px rgba(0,212,255,0.4)';
      };
      btn.onmouseleave = () => {
        btn.style.borderColor = '#444';
        btn.style.boxShadow = 'none';
      };

      btn.onclick = () => {
        buttonWrapper.querySelectorAll('button').forEach(b => b.disabled = true);
        const isCorrect = ans.correct;

        if (isCorrect) {
          btn.style.background = 'rgba(16,185,129,0.3)';
          btn.style.borderColor = '#10b981';
          btn.style.color = '#10b981';
          QuestFileStructure.gameScore.points += 25;
          QuestFileStructure.gameScore.quizzesCompleted += 1;
          setTimeout(() => {
            overlay.remove();
            QuestFileStructure.showToast(`✓ Correct! +25 pts • Total: ${QuestFileStructure.gameScore.points} pts`, '#10b981');
            onComplete(true);
          }, 800);
        } else {
          btn.style.background = 'rgba(239,68,68,0.3)';
          btn.style.borderColor = '#ef4444';
          btn.style.color = '#ef4444';

          answers.forEach(a => {
            const buttons = Array.from(buttonWrapper.querySelectorAll('button'));
            if (a.correct && a.text !== ans.text) {
              const correctBtn = buttons.find(b => b.textContent === a.text);
              if (correctBtn) {
                correctBtn.style.background = 'rgba(16,185,129,0.2)';
                correctBtn.style.borderColor = '#10b981';
                correctBtn.style.color = '#10b981';
              }
            }
          });

          setTimeout(() => {
            overlay.remove();
            QuestFileStructure.showToast(`✗ Wrong. Correct: ${quizData.options.find(o => o.correct).text}`, '#ef4444');
            onComplete(false);
          }, 1500);
        }
      };

      buttonWrapper.appendChild(btn);
    });

    card.appendChild(title);
    card.appendChild(question);
    card.appendChild(buttonWrapper);
    overlay.appendChild(card);
    document.body.appendChild(overlay);
  }

  // ─── Building Factory ───────────────────────────────────────────────────
  static makeBuilding(id, x, buildingName, lessonHTML, quizQuestion) {
    let visitedThisSession = false;

    return {
      id,
      x,
      buildingName,
      lesson: lessonHTML,
      quiz: quizQuestion,
      visitedThisSession,
      interact: async function() {
        if (this.visitedThisSession) {
          QuestFileStructure.showToast(`You've already cleared ${buildingName}!`, '#64748b');
          return;
        }
        QuestFileStructure.gameScore.buildingsVisited += 1;
        this.visitedThisSession = true;
        await QuestFileStructure.showLessonAndQuiz(buildingName, lessonHTML, quizQuestion);

        if (QuestFileStructure.gameScore.buildingsVisited === 6) {
          setTimeout(() => {
            QuestFileStructure.showToast(`🎉 Quest Complete! Final Score: ${QuestFileStructure.gameScore.points} pts`, '#f59e0b');
          }, 500);
        }
      }
    };
  }

  // ─── Constructor ─────────────────────────────────────────────────────────
  constructor(gameEnv) {
    const path = gameEnv.path;
    const width = gameEnv.innerWidth;
    const height = gameEnv.innerHeight;

    // Score display overlay
    const scoreDisplay = document.createElement('div');
    scoreDisplay.id = 'game-score-display';
    scoreDisplay.style.cssText = `
      position: fixed; top: 10px; left: 10px; z-index: 500;
      background: rgba(26, 26, 46, 0.9); border: 2px solid #00d4ff;
      color: #e2e8f0; padding: 12px 20px; border-radius: 6px;
      font-family: 'Courier New', monospace; font-size: 14px;
      font-weight: bold;
    `;
    scoreDisplay.textContent = `Score: ${QuestFileStructure.gameScore.points} | Buildings: ${QuestFileStructure.gameScore.buildingsVisited}/6`;

    if (!document.getElementById('game-score-display')) {
      document.body.appendChild(scoreDisplay);
    }

    // Update score display every second
    const updateScoreDisplay = () => {
      if (document.getElementById('game-score-display')) {
        document.getElementById('game-score-display').textContent = 
          `Score: ${QuestFileStructure.gameScore.points} | Buildings: ${QuestFileStructure.gameScore.buildingsVisited}/6 | Quizzes: ${QuestFileStructure.gameScore.quizzesCompleted}/6`;
      }
    };
    setInterval(updateScoreDisplay, 500);

    // Create the 6 buildings with lessons and quizzes
    const buildings = [
      QuestFileStructure.makeBuilding(
        'src_building',
        width * 0.15,
        '📁 The src/ Folder',
        `<div><strong>The src/ folder is your project\'s heart.</strong> All source code lives here: JavaScript files, Python scripts, Java classes — whatever language you use.</div>
         <div style="margin-top: 12px; padding: 12px; background: #292945; border-left: 3px solid #00d4ff;">
         <code style="color: #a6e3a1;">my-project/<br/>├── src/<br/>│   ├── main.js<br/>│   ├── utils.js<br/>│   └── components/<br/>├── tests/<br/>└── README.md</code>
         </div>
         <div style="margin-top: 12px;"><strong>Why?</strong> Separates code from config, tests, and docs. Makes automation easier. Looks professional.</div>`,
        {
          question: 'Where should your main application code go?',
          options: [
            { text: 'A) Random files in the root directory', correct: false },
            { text: 'B) Inside the src/ folder', correct: true },
            { text: 'C) Hide it in node_modules', correct: false },
            { text: 'D) Spread across multiple folders with no structure', correct: false }
          ]
        }
      ),

      QuestFileStructure.makeBuilding(
        'tests_building',
        width * 0.35,
        '🧪 The tests/ Folder',
        `<div><strong>Tests verify your code works.</strong> They live in a separate folder so they don't clutter src/ and are easy to run together.</div>
         <div style="margin-top: 12px; padding: 12px; background: #292945; border-left: 3px solid #00d4ff;">
         <code style="color: #a6e3a1;">tests/<br/>├── test_main.js<br/>├── test_utils.js<br/>└── fixtures/</code>
         </div>
         <div style="margin-top: 12px;"><strong>Best practice:</strong> Mirror your src/ structure. If src/utils.js exists, create tests/test_utils.js with the same functions.</div>`,
        {
          question: 'Why keep tests separate from src/?',
          options: [
            { text: 'A) Tests take up too much space', correct: false },
            { text: 'B) To keep src/ clean and run tests independently', correct: true },
            { text: 'C) So programmers can\'t find them', correct: false },
            { text: 'D) Tests are optional and don\'t matter', correct: false }
          ]
        }
      ),

      QuestFileStructure.makeBuilding(
        'config_building',
        width * 0.5,
        '⚙️ configs/ & .env',
        `<div><strong>Configuration files tell your app how to behave</strong> in different environments (dev, testing, production).</div>
         <div style="margin-top: 12px; padding: 12px; background: #292945; border-left: 3px solid #00d4ff;">
         <code style="color: #a6e3a1;">configs/<br/>  ├── dev.json<br/>  ├── prod.json<br/>  └── test.json<br/>.env (NEVER commit this!)</code>
         </div>
         <div style="margin-top: 12px;"><strong>⚠️ Important:</strong> NEVER put secrets in .env then commit it. Add .env to .gitignore. Secrets belong in environment variables or a secrets manager.</div>`,
        {
          question: 'What should NEVER be committed to Git?',
          options: [
            { text: 'A) The .env file with API keys and secrets', correct: true },
            { text: 'B) Your source code (src/ folder)', correct: false },
            { text: 'C) Configuration files like config.json', correct: false },
            { text: 'D) Your README.md', correct: false }
          ]
        }
      ),

      QuestFileStructure.makeBuilding(
        'docs_building',
        width * 0.65,
        '📖 docs/ Folder',
        `<div><strong>Documentation lives in docs/.</strong> Not just a README—include API docs, architecture diagrams, setup guides, troubleshooting, etc.</div>
         <div style="margin-top: 12px; padding: 12px; background: #292945; border-left: 3px solid #00d4ff;">
         <code style="color: #a6e3a1;">docs/<br/>├── README.md<br/>├── API.md<br/>├── SETUP.md<br/>├── ARCHITECTURE.md<br/>└── images/</code>
         </div>
         <div style="margin-top: 12px;"><strong>Good documentation = happy teammates.</strong> Someone should be able to run your project without asking you questions.</div>`,
        {
          question: 'What is a good reason for detailed documentation?',
          options: [
            { text: 'A) To show off (nobody reads it anyway)', correct: false },
            { text: 'B) So teammates can set up and contribute without constant questions', correct: true },
            { text: 'C) It\'s required by law', correct: false },
            { text: 'D) To waste time instead of coding', correct: false }
          ]
        }
      ),

      QuestFileStructure.makeBuilding(
        'deps_building',
        width * 0.78,
        '📦 Dependencies',
        `<div><strong>Dependencies list is crucial.</strong> Every language has a standard: package.json (Node), requirements.txt (Python), pom.xml (Java).</div>
         <div style="margin-top: 12px; padding: 12px; background: #292945; border-left: 3px solid #00d4ff;">
         <code style="color: #a6e3a1;"># package.json<br/>{ "dependencies": { "express": "^4.18.0" } }<br/><br/># requirements.txt<br/>flask==2.0.1<br/>requests==2.28.0</code>
         </div>
         <div style="margin-top: 12px;"><strong>NEVER commit node_modules/.</strong> It's huge. Instead, commit the lock file (package-lock.json or requirements.txt) so teammates can run install.</div>`,
        {
          question: 'Why shouldn\'t you commit node_modules/?',
          options: [
            { text: 'A) It\'s bad luck', correct: false },
            { text: 'B) It\'s massive (thousands of files) and recreatable with npm install', correct: true },
            { text: 'C) Modules will break if you commit them', correct: false },
            { text: 'D) Only Node devs use it', correct: false }
          ]
        }
      ),

      QuestFileStructure.makeBuilding(
        'gitignore_building',
        width * 0.88,
        '🚫 .gitignore',
        `<div><strong>.gitignore tells Git what files to ignore.</strong> It's a text file listing patterns of files never to track.</div>
         <div style="margin-top: 12px; padding: 12px; background: #292945; border-left: 3px solid #00d4ff;">
         <code style="color: #a6e3a1;"># .gitignore<br/>node_modules/<br/>.env<br/>dist/<br/>*.log<br/>.DS_Store<br/>__pycache__/</code>
         </div>
         <div style="margin-top: 12px;"><strong>Pro tip:</strong> Use gitignore.io or your language's standard template. Never manually hunt for what to ignore.</div>`,
        {
          question: 'Which files should be in .gitignore?',
          options: [
            { text: 'A) Your README and docs', correct: false },
            { text: 'B) node_modules/, .env, build artifacts, and temporary logs', correct: true },
            { text: 'C) Everything to keep the repo empty', correct: false },
            { text: 'D) Only really large files', correct: false }
          ]
        }
      ),
    ];

    // Player sprite
    const playerData = {
      id: 'DevHero',
      src: path + '/images/gamify/chillguy.png',
      SCALE_FACTOR: 5,
      STEP_FACTOR: 1000,
      ANIMATION_RATE: 50,
      INIT_POSITION: { x: 50, y: height / 2 },
      pixels: { height: 512, width: 384 },
      orientation: { rows: 4, columns: 3 },
      down: { row: 0, start: 0, columns: 3 },
      downRight: { row: 1, start: 0, columns: 3, rotate: Math.PI / 16 },
      downLeft: { row: 2, start: 0, columns: 3, rotate: -Math.PI / 16 },
      right: { row: 1, start: 0, columns: 3 },
      left: { row: 2, start: 0, columns: 3 },
      up: { row: 3, start: 0, columns: 3 },
      upRight: { row: 1, start: 0, columns: 3, rotate: -Math.PI / 16 },
      upLeft: { row: 2, start: 0, columns: 3, rotate: Math.PI / 16 },
      hitbox: { widthPercentage: 0.45, heightPercentage: 0.2 },
      keypress: { up: 87, left: 65, down: 83, right: 68 }
    };

    // Collision detection for buildings
    const checkBuildingCollision = () => {
      buildings.forEach(building => {
        const minDist = width * 0.08;
        const player = document.getElementById('DevHero_id');
        
        if (player) {
          const playerX = parseFloat(player.style.left || 0) + 50;
          const playerY = parseFloat(player.style.top || 0) + 50;
          
          const dist = Math.sqrt(
            Math.pow(playerX - building.x, 2) + 
            Math.pow(playerY - height / 2, 2)
          );
          
          if (dist < minDist) {
            QuestFileStructure.showToast(`Press E to enter ${building.buildingName}`, '#00d4ff');
          }
        }
      });
    };

    setInterval(checkBuildingCollision, 200);

    // E key handler for building interaction
    document.addEventListener('keydown', (e) => {
      if (e.key.toLowerCase() === 'e') {
        const player = document.getElementById('DevHero_id');
        if (player) {
          const playerX = parseFloat(player.style.left || 0) + 50;
          const playerY = parseFloat(player.style.top || 0) + 50;
          
          buildings.forEach(building => {
            const dist = Math.sqrt(
              Math.pow(playerX - building.x, 2) + 
              Math.pow(playerY - height / 2, 2)
            );
            
            if (dist < width * 0.1) {
              building.interact();
            }
          });
        }
      }
    }, { once: false });

    // Game entities
    this.classes = [
      { class: GameEnvBackground, data: { name: 'quest1_bg', src: path + '/images/gamify/siliconvalley.png' } },
      { class: Player, data: playerData },
    ];

    // Create NPCs at each building location
    buildings.forEach(building => {
      this.classes.push({
        class: NPC,
        data: {
          id: building.id,
          src: path + '/images/gamify/robot.png',
          SCALE_FACTOR: 6,
          ANIMATION_RATE: 100,
          INIT_POSITION: { x: building.x, y: height / 2 },
          pixels: { height: 256, width: 256 },
          orientation: { rows: 1, columns: 1 },
          down: { row: 0, start: 0, columns: 1 },
          hitbox: { widthPercentage: 0.5, heightPercentage: 0.5 },
          greeting: `${building.buildingName} — Press E to learn!`,
          reaction: function() {
            if (this.dialogueSystem) this.showReactionDialogue();
          },
          interact: async function() {
            await building.interact();
          }
        }
      });
    });
  }
}

export const gameLevelClasses = [QuestFileStructure];
export { GameControl };
````

## Quest Design Guide

Each quest follows the same pattern:

- **Background**: Sets the visual scene for the quest
- **Player**: The Dev Hero the student controls with WASD
- **NPCs**: Characters that represent concepts (files, errors, targets) and deliver dialogue when approached
- **Dialogues**: Step through lesson content by walking up to NPCs and pressing the interact key

### Adding a New Quest

1. Create a new `%%js` code cell
2. Add `// GAME_RUNNER: <your challenge description>`
3. Define a new level class with the concept NPCs
4. Export `gameLevelClasses` and `GameControl`
5. Run `make` to convert and test on the web

### Scoring Ideas

Embed point values in NPC dialogue lines so students know what each interaction is worth:

```javascript
dialogues: [
  'I am the Makefile!',
  'Objective complete: +80 pts'
]
```
