## Introduction

Picture this: You're getting ready for school in the morning, so first, you check the weather forecast. 

**IF** you see that it's raining, you grab your umbrella. **BUT THEN** you ask yourself another question: "How cold is it?" 
    - **IF** it's also cold, you pack BOTH your umbrella AND your jacket
    - **ELSE** (if it's warm but rainy), you just take the umbrella

**ELSE** (if it's NOT raining), you skip the umbrella. **BUT** you still need to check: "Is it cold outside?"
    - **IF** it's cold, you take just your jacket
    - **ELSE** (if it's warm and not raining), you're good to go with nothing!

Notice the temperature check in both the **IF** and the **ELSE** part of the statement? That's the nested part, because nested basically means embedded in this scenario.

## Syntax

Since the nested conditional is an embedded if statement, there needs to be an outer if statement too. You can see the formatting like so:



```python
%%js

// Check the weather before leaving for school
if (raining) {
    // First decision: It's raining, so take umbrella
    takeUmbrella();
    
    // Nested decision INSIDE the raining condition
    if (cold) {
        takeJacket();
        console.log("Taking umbrella AND jacket");
    } else {
        console.log("Taking just umbrella");
    }
    
} else {
    // First decision: Not raining, skip umbrella
    
    // Nested decision INSIDE the not-raining condition
    if (cold) {
        takeJacket();
        console.log("Taking just jacket");
    } else {
        console.log("Nothing needed - good to go!");
    }
}
```

**Notice how it's just like the example we had above, except now you have to embed the condition of being cold inside the `{}` brackets. Also, console.log just prints the output in the browser console tab.**
