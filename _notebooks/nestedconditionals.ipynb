```python
from IPython.display import HTML, display

display(HTML('''
<!DOCTYPE html>
<html>
<head>
    <style>
        * { box-sizing: border-box; margin: 0; padding: 0; }
        .container { font-family: Arial, sans-serif; max-width: 1200px; margin: 0 auto; }
        .header { background: #2d3748; color: white; padding: 20px; }
        .editor-section { display: flex; height: 500px; }
        .editor-panel, .output-panel { flex: 1; display: flex; flex-direction: column; }
        .panel-header { background: #4a5568; color: white; padding: 10px; font-weight: bold; }
        textarea { flex: 1; padding: 15px; font-family: monospace; font-size: 14px; border: none; background: #1a202c; color: #e2e8f0; resize: none; }
        .output { flex: 1; padding: 15px; background: #0d1117; color: #0dff0d; font-family: monospace; font-size: 14px; overflow: auto; white-space: pre-wrap; }
        .error { color: #ff6b6b; }
        .controls { padding: 10px; background: #edf2f7; display: flex; gap: 10px; }
        button { padding: 10px 20px; border: none; border-radius: 5px; cursor: pointer; font-weight: bold; }
        .run-btn { background: #48bb78; color: white; }
        .run-btn:hover { background: #38a169; }
        .reset-btn { background: #cbd5e0; }
        .reset-btn:hover { background: #a0aec0; }
    </style>
</head>
<body>
    <div class="container">
        <div class="header">
            <h2>JavaScript Code Runner</h2>
            <p>Write and execute JavaScript code in real-time</p>
        </div>
        
        <div class="controls">
            <button class="run-btn" onclick="runCode()">▶ Run Code</button>
            <button class="reset-btn" onclick="resetCode()">↻ Reset</button>
        </div>
        
        <div class="editor-section">
            <div class="editor-panel">
                <div class="panel-header">Code Editor</div>
                <textarea id="codeEditor">// Write your JavaScript code here
console.log("Hello, World!");

// Variables and operations
const x = 10;
const y = 5;
console.log("x + y =", x + y);

// Functions
function greet(name) {
    return "Hello, " + name + "!";
}

console.log(greet("Claude"));

// You can return a value
return "Code executed successfully!";</textarea>
            </div>
            
            <div class="output-panel">
                <div class="panel-header">Output</div>
                <div id="output" class="output">Click "Run Code" to see output...</div>
            </div>
        </div>
    </div>
    
    <script>
        const defaultCode = document.getElementById('codeEditor').value;
        
        function runCode() {
            const code = document.getElementById('codeEditor').value;
            const outputDiv = document.getElementById('output');
            outputDiv.innerHTML = '';
            outputDiv.className = 'output';
            
            const logs = [];
            const oldLog = console.log;
            const oldError = console.error;
            const oldWarn = console.warn;
            
            console.log = (...args) => {
                logs.push(args.map(arg => 
                    typeof arg === 'object' ? JSON.stringify(arg, null, 2) : String(arg)
                ).join(' '));
            };
            
            console.error = (...args) => {
                logs.push('ERROR: ' + args.join(' '));
            };
            
            console.warn = (...args) => {
                logs.push('WARNING: ' + args.join(' '));
            };
            
            try {
                const result = new Function(code)();
                
                if (result !== undefined) {
                    logs.push('\\n→ Returned: ' + (typeof result === 'object' ? JSON.stringify(result, null, 2) : String(result)));
                }
                
                outputDiv.textContent = logs.join('\\n') || '✓ Code executed (no output)';
            } catch (error) {
                outputDiv.className = 'output error';
                outputDiv.textContent = '❌ Error:\\n' + error.toString();
            } finally {
                console.log = oldLog;
                console.error = oldError;
                console.warn = oldWarn;
            }
        }
        
        function resetCode() {
            document.getElementById('codeEditor').value = defaultCode;
            document.getElementById('output').textContent = 'Click "Run Code" to see output...';
            document.getElementById('output').className = 'output';
        }
    </script>
</body>
</html>
'''))
```
