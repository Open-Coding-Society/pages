### Game Coding

As teams of 5 or 6 we will be working on a personal game. We will be changing, adding, or extending levels in the mansion or mystery game.

- Each team will make a team repository, using OCS pages as a template. 

- Each team of 6 will own, update, and customize multiple game levels.

- Each team will establish a story and determine how the game ends.

- As pairs of 2 you will prepare final review on how you planned, coded, and prepared for contributions and live reviews during Sprint 4.

- Each team will prepare a final review.
