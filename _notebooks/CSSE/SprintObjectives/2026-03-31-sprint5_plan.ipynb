## RPG Game in Depth

As teams of 3 across the class will all be working on the same RPG Game. We will be adding or extending the system by enhancing Game Classes and Objects.. 

- Each team of 3 will own, update, and customize a topic(s) within the GameEngine.

- Each team will perform a review and complete 3 or more pull requests to OCS system.

- Each team will prepare a blog that and Game Runners to describe the game enhancements they designed and coded.

- Team will share 1 or more of their accomplishments with entire class.

## Materials

Many materials have been provided for key topics.  These should be considered as starters or guidelines.

- Key Objectives
  - Read the materials provided on pages.
  - Learn your topic in the game.
  - Build ideas on how to change the game.
  - Add learnings and ideas to a blog and category for the Search system.
  - Think that you will be leaving a legacy of work to the next RPG coders, pay forward.

- Selection
  - Pick your 1st topic, 2nd choice, 3rd choice.
  - The two CSSSE classrooms will have a coordinated signup.
  - After selection you will perform a PR on pages for your topic(s), with team members names.

## Checklist Suggestions
Here are suggested activities.  These should be considered minmal, but can be enhanced if they lack detail.

- Scrum Master of Scrum Masters (Torin)
  - Start Issue of Team Assignments
  - Starts Project - Kanban Board and Timeline using Issues from individuals
  - Make sure we are not duplicating by doing reviews with each Scrum Master and Assistant Scrum Master
  - Accept pull requests and ensure they match Team and Individual Assignments
  - Pull in other Scrum Master to support integrations.
- Scrum Master
  - Build an Outline of Topic and Lesson Materials using provided materials
  - Discuss Ouline of Topic and of Lesson Materias with Team and with Scrum Master of Scrum Masters
  - Start Individual Issue(s) for Team Members, be sure Issue(s) have burn down lists.
  - Make sure your Issue(s) get on to the Kanban Board
- Teams of 3
  - Team meets and describes parts of game that will have updates and lessons, capture ideas in Team Issue.
  - Break down parts of game into Individual Issue(s) for each Developer
  - Each Developer build skeletal notebooks into game, be sure to update front matter. 
  - Make sure skeletal notebooks get into game in 1st week of sprint
  - Make second pass on notebooks to update content and future ideas for system.  Make this interesting including game runner code fragments that run deployed.  Updating DOM and having help interactive is highly encouraged.
  - Team makes pull request.
