### **Introduction to Mathematical Expresions**

ðŸ“Œ What are Mathematical Expressions in Programming?
In programming, a mathematical expression is a combination of values(numbers), variables and operators that the program evluates to give a numeric result. 

For Example:

`5 + 3 * 2;`

> **Key Takeaway:** As you can see, mathematical expressions in JS are practically identical to mathematical expressions found in typical math.

---

### **Key Mathematical Operators**

| Operator         | Symbol | Example     | Result | Example use case       |
|------------------|--------|-------------|--------|-----------------------------|
| Addition         | `+`    | `2 + 3`     | `5`    | `console.log(2 + 3);`       |
| Subtraction      | `-`    | `5 - 2`     | `3`    | `console.log(5 - 2);`       |
| Multiplication   | `*`    | `4 * 3`     | `12`   | `console.log(4 * 3);`       |
| Division         | `/`    | `10 / 2`    | `5`    | `console.log(10 / 2);`      |
| Modulo        | `%` | `7 % 3`     | `1`    | `console.log(7 % 3);`       |
| Exponentiation   | `**`   | `2 ** 3`    | `8`    | `console.log(2 ** 3);`      |

You can also use inequality operators in if statements (>, <, ==, <=, =>)

---

### **Variables in Expressions**
You can use variables to store values and build expressions:

### **Order of Operations**
Just like in typical math, JavaScript follows PEMDAS:
- Parentheses
- Exponents
- Multiplication/Division (left to right)
- Addition/Subtraction (left to right)

For example:

`let result = (4 + 2) * 3; // (4+2) * 3 = 6 * 3 = 18`

```python
%%js

// CODE_RUNNER: Fix the assignment error in variable declaration 

x = 5; // Error: x is not defined

console.log(x);
```

```python
%%js

// CODE_RUNNER: Variables in Expressions and Order of Operations illustration

let x = 10;
let y = 3;
let total1 = x + y * 2; // total = 10 + (3 Ã— 2) = 16
let total2 = (x + y) * 2; // total2 = (10 + 3) Ã— 2 = 26

console.log("The total is: " + total1); // Following PEMDAS
console.log("The total2 is: " + total2); // Overriding PEMDAS with parentheses
```

## Advanced Example

The below code cells shows the following advanced ways of building a custom way to run code. This uses... 

- Style for textarea and buttons
- HTML Document Object Mode (DOM) definitions
- Event handling
- Evaluation of the textarea example code

```python
%%html

<!-- Penguins custom code runner using an HTML cell, leave in lesson to give credit -->
<style>
  /* Basic styling for javascript */
  textarea { width: 100%; height: 200px; }
  button { margin-top: 10px; padding: 8px 16px; font-size: 14px; cursor: pointer; }
</style>

<!-- Basic HTML structure for the challenge -->
<h2>JavaScript - Math Challenge user conventional Style, HTML, and JS</h2>
<p>Set a variable <code>x</code> to any number. Then calculate:</p>
<ul>
  <li><code>x + 5</code></li>
  <li><code>x * 2</code></li>
  <li><code>x % 3</code> (remainder when x is divided by 3)</li>
</ul>
<p>Print all three results in a single line. Try different values of <code>x</code>!</p>

<!-- Code editor for user input -->
<textarea id="code">
// Example: change the value of x and run the code
let x = 4;
let addResult = x + 5;
let multiplyResult = x * 2;
let moduloResult = x % 3;

document.getElementById('output').textContent = addResult + ' ' + multiplyResult + ' ' + moduloResult;
</textarea>
<br>
<!--Run code action button-->
<button id="run-button">Run Code</button>
<h3>Output:</h3>
<div id="output"></div>

<script>
  // Attach event listener in JS to the run button 
  document.getElementById('run-button').addEventListener('click', function() {
    const code = document.getElementById('code').value;
    const outputDiv = document.getElementById('output');
    outputDiv.textContent = '';
    try {
      eval(code);  // execute student code
    } catch (err) {
      outputDiv.textContent = 'Error: ' + err.message;
    }
  });
</script>
```
