  <h2>Detecting when a player is close</h2>
  <p class="lead">Every frame, <code>_findNearbyEnemy()</code> loops every living enemy and measures the straight-line distance to the player using <em>Euclidean distance</em> (the Pythagorean theorem). If it's below 120 px, a collision is flagged.</p>

  <div class="anno-wrap">
    <div class="anno-code">
      <div class="code-header"><span class="code-title">_findNearbyEnemy() · the distance check</span><span class="code-tag">JS</span></div>
      <div class="code-body"><pre><span class="fn">_findNearbyEnemy</span>(threshold = <span class="num">120</span>) {
  <span class="kw">const</span> pos = <span class="kw">this</span>.<span class="fn">_playerLogicalPos</span>();
  <span class="kw">if</span> (!pos) <span class="kw">return null</span>;
  <span class="kw">let</span> closest = <span class="kw">null</span>, best = threshold;

  <span class="kw">for</span> (<span class="kw">const</span> e <span class="kw">of</span> <span class="kw">this</span>.<span class="prop">_worldEnemies</span>) {
    <span class="kw">if</span> (e.<span class="prop">defeated</span>) <span class="kw">continue</span>;        <span class="cmt">// A — skip dead enemies</span>
    <span class="kw">const</span> d = Math.<span class="fn">hypot</span>(            <span class="cmt">// B — √(dx² + dy²)</span>
      pos.<span class="prop">x</span> - e.<span class="prop">logicalX</span>,
      pos.<span class="prop">y</span> - e.<span class="prop">logicalY</span>
    );
    <span class="kw">if</span> (d &lt; best) { best = d; closest = e; } <span class="cmt">// C — keep nearest</span>
  }
  <span class="kw">return</span> closest; <span class="cmt">// null if nobody within 120px</span>
}</pre></div>
    </div>
    <div class="anno-notes">
      <div class="anno-note"><div class="anno-icon">A</div><div class="anno-text"><strong>e.defeated check</strong> — once an enemy is flagged as dead, the loop simply skips them. No distance math wasted on enemies that no longer exist.</div></div>
      <div class="anno-note"><div class="anno-icon">B</div><div class="anno-text"><strong>Math.hypot(dx, dy)</strong> is shorthand for √(dx² + dy²) — the length of the hypotenuse. It gives the true straight-line distance between two points (Pythagorean theorem in 2D).</div></div>
      <div class="anno-note"><div class="anno-icon">C</div><div class="anno-text"><strong>Only the nearest enemy is returned.</strong> <code>best</code> shrinks each iteration — so the threshold doubles as a "distance so far" tracker, not just a cutoff.</div></div>
    </div>
  </div>
