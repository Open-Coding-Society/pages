# Robot Logic Game

Six levels. One robot. A small set of instructions.

Your job is to drag blocks into the **Program** panel and press **Run** to make the robot reach the green goal tile. Every level reads stars based on how *few* blocks your program uses — getting to the goal is enough to win, but a tighter program earns more.

You'll meet three control patterns along the way:

1. **Sequence** — instructions run top to bottom, one after another.
2. **Iteration** — `Repeat N times` runs a group of blocks N times.
3. **Selection** — `If Can Move Forward` runs a group of blocks only when the path ahead is clear.

That's the whole language. Levels just rearrange those three ideas into bigger and bigger puzzles.

## The blocks

| Block | What it does |
| --- | --- |
| `Move Forward` | Move one tile in the direction the robot is facing. Blocked by walls and grid edges. |
| `Move Backward` | Move one tile *behind* the robot, without turning. |
| `Turn Left` | Rotate 90° counter-clockwise. |
| `Turn Right` | Rotate 90° clockwise. |
| `Repeat N times` | Loop — run the body N times. The N is editable in the block. |
| `If Can Move Forward` | Run the body only if the path ahead is clear. |

Reading a program: blocks run top-to-bottom. Blocks *inside* a `Repeat` or `If` body count toward your block total — so `Repeat 6 [Forward]` is 2 blocks, not 7.

## Level 1 — First Steps

**Concept: Sequence.** Instructions run in the order you stack them. The robot reads top to bottom and does each block one at a time. Order is meaning — `Forward, Forward, Turn Left` is a different path from `Turn Left, Forward, Forward`.

The level introduces the `Repeat N times` block. You don't *need* it to win, but using it earns more stars.

**Tips**

- Start at column 0 row 3, facing East. The goal is at column 6 row 0 (top-right).
- Move East six tiles, turn left, move three more — that reaches the goal.
- The brute-force solution stacks 10 blocks. Wrap the runs of `Forward` in `Repeat` to cut that down.
- Optimal: `Repeat 6 [Forward]` + `Turn Left` + `Repeat 3 [Forward]` = **5 blocks**.
- **Star thresholds:** ≤5 blocks → ★★★, ≤8 blocks → ★★, otherwise ★.

## Level 2 — Around the Wall

**Concept: Sequence with planning.** Same instruction set as Level 1, but now there's a wall in the way. The robot can't move into a wall — it just stops, blocked. So you can't barrel forward; you have to plan a detour before pressing Run.

Walking-into-walls is the most common bug at this stage. Trace the path tile by tile in your head before you Run.

**Tips**

- Walls are at column 3, rows 2/3/4 — a vertical line directly between you and the goal.
- Route around the wall: go up (North) or down (South) past the wall, then East across, then back to row 3.
- Both detours work. The shortest takes 11 blocks.
- A clean three-star solution: `Repeat 2 [Forward]` + `Turn Left` + `Repeat 2 [Forward]` + `Turn Right` + `Repeat 4 [Forward]` + `Turn Right` + `Repeat 2 [Forward]`.
- **Star thresholds:** ≤11 blocks → ★★★, ≤14 blocks → ★★.

## Level 3 — The Long Way

**Concept: Iteration.** When the same block repeats over and over, wrap it in `Repeat N times`. The body inside a `Repeat` block runs N times, then the program continues to the next block.

```
Repeat 6 times:
    Forward
```

Six lines of `Forward` collapse into one `Repeat` plus one `Forward` inside — two blocks instead of six. This is the single biggest source of star count compression in the game.

**Tips**

- The wall is a long vertical line at column 3 — much longer than Level 2.
- Without `Repeat`, you'd need ~19 blocks. With nested-free `Repeat`s for each long run of forwards, the optimal is **11 blocks**.
- A `Repeat` block is one block — and the blocks *inside* it count toward your total. So `Repeat 6 [Forward]` is 2 blocks, not 7.
- Start by writing the brute-force version. Then look for runs of identical blocks and replace each run with a `Repeat`.
- **Star thresholds:** ≤11 blocks → ★★★, ≤17 blocks → ★★.

## Level 4 — Reverse Course

**Concept: A new block.** `Move Backward` moves the robot one tile *behind* its facing — without turning around. Sometimes the goal is behind you, and turning 180° to face it costs two `Turn` blocks. `Move Backward` skips that cost entirely.

Pick the right tool. `Forward` and `Backward` look symmetric, but in a level where you're already facing away from the goal, `Backward` saves blocks.

**Tips**

- The robot starts at column 5, facing East — but the goal is at column 0, *behind* it.
- A naive solution: turn 180° (`Turn Left, Turn Left`), then `Repeat 5 [Forward]`. That's 3 blocks.
- A backward solution: `Repeat 5 [Backward]`. That's **2 blocks** — three stars.
- **Star thresholds:** ≤2 blocks → ★★★, ≤4 blocks → ★★.

## Level 5 — Corridor Maze

**Concept: Putting it together.** No new blocks. This level forces you to combine sequence, `Repeat`, and the backward block into a single program. You'll see *nested* `Repeat`s shine here — a small repeating pattern (e.g. "walk along a corridor, turn, drop down a row") wrapped inside a larger `Repeat` that repeats the whole pattern several times.

```
Repeat 3 times:
    Repeat 5 times:
        Forward
    Turn Right
    Forward
    Turn Right
```

Read it aloud: *"Three times: walk five tiles, turn right, drop one tile, turn right."*

**Tips**

- Three horizontal walls separate the grid into four corridors. You must zigzag through them.
- The corridors alternate direction — across, down a row, back across the other way, down again.
- Optimal solutions use a `Repeat` with a nested pattern. Aim for 13 blocks.
- This is the hardest level so far — expect to need 20+ blocks on the first attempt and shrink from there.
- **Star thresholds:** ≤13 blocks → ★★★, ≤20 blocks → ★★.

## Level 6 — Right-Hand Spiral

**Concept: Selection.** `If Can Move Forward` runs its body *only* when the path ahead is clear. Drop it on its own and it does almost nothing useful — *"move if you can"* on a single tile. The pattern gets powerful when you put it inside a `Repeat`.

```
Repeat 3 times:
    Repeat 10 times:
        If Can Move Forward:
            Forward
    Turn Right
```

Read it aloud: *"Three times: walk until blocked, then turn right."* That's the **right-hand rule** — a classic maze-solving trick. The robot doesn't know how long each corridor is; it just *tries* to move 10 times and the `If` quietly skips the move when it would crash. Your code adapts to the world instead of assuming it.

**Tips**

- This is the only level with the `If Can Move Forward` block available. Use it.
- Optimal solution: **5 blocks total** — exactly the pattern shown in the concept above.
- The outer `Repeat 3` runs the walk-then-turn three times; the inner `Repeat 10` is just a generous upper bound for "until blocked."
- Without `If Can Move Forward`, you'd have to count corridor lengths by hand and write each leg separately — much longer.
- **Star thresholds:** ≤5 blocks → ★★★, ≤9 blocks → ★★.

## Recap

| Pattern | Block | When to use |
| --- | --- | --- |
| Sequence | stacked blocks | always — it's the default |
| Iteration | `Repeat N times` | doing the same thing N times |
| Selection | `If Can Move Forward` | only doing something *if* the path is clear |

Every level can be solved with combinations of these three — and a good program is a *short* program.

## Hacks (Challenges)

1. **3-Star Sweep.** Earn three stars on every level. Stars are saved on this device.
2. **Manual Mode.** Solve Level 6 *without* using `If Can Move Forward`. How many more blocks did you need? Why?
3. **One-Block Diet.** Pick any 3-star solve and try to remove one more block. If you can't, explain in plain English why the program is already minimal.
4. **Design Your Own Level.** Open `_projects/robot-logic-game/levels/levelConfigs.js` and add a new entry. Try a 9×9 grid, a few wall clusters, and tune `starThresholds` so the optimal solution earns 3 stars.

## Reflection

1. **Compression.** Which block (`Repeat`, `If Can Move Forward`, `Backward`) gave you the biggest reduction in block count? On which level?
2. **Trace.** Pick one of your three-star programs. On paper, walk through it block-by-block: where is the robot after each block runs? Which blocks do nothing useful?
3. **Generalization.** `Repeat N times` always runs *exactly* N times. What kind of block would you want for "keep going *while* something is true and stop when it isn't"? Sketch how it would look.
