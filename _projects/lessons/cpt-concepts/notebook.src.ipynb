<script>
pdfjsLib.GlobalWorkerOptions.workerSrc =
  'https://cdnjs.cloudflare.com/ajax/libs/pdf.js/3.11.174/pdf.worker.min.js';

(function() {
  const $ = id => document.getElementById(id);

  // API base URL — matches the site's config pattern
  const javaURI = (location.hostname === 'localhost' || location.hostname === '127.0.0.1')
    ? 'http://localhost:8585' : 'https://spring.opencodingsociety.com';

  // Preset FRQ PDFs
  const PRESET_YEARS = [2025,2024,2023,2022,2021,2019,2018,2017,2016,2015,2014,2013,2011,2010,2009,2008,2007,2006,2005];
  function presetPdfUrl(year) {
    const base = document.querySelector('base');
    const prefix = base ? base.href.replace(/\/$/, '') : '';
    return prefix + '/assets/pdfs/frq/ap-csa-frq-' + year + '.pdf';
  }
  function refSheetUrl() {
    const base = document.querySelector('base');
    const prefix = base ? base.href.replace(/\/$/, '') : '';
    return prefix + '/assets/pdfs/reference/ap-csa-java-quick-reference.pdf';
  }

  function timeLimitMinutes(year) {
    if (!year) return 90;
    return year < 2020 ? 105 : 90;
  }
  function formatTime(seconds) {
    if (seconds < 0) seconds = 0;
    const h = Math.floor(seconds / 3600);
    const m = Math.floor((seconds % 3600) / 60);
    const s = seconds % 60;
    if (h > 0) return h + ':' + String(m).padStart(2,'0') + ':' + String(s).padStart(2,'0');
    return String(m).padStart(2,'0') + ':' + String(s).padStart(2,'0');
  }
  function escapeHtml(s) {
    return s.replace(/[&<>"']/g, c => ({'&':'&amp;','<':'&lt;','>':'&gt;','"':'&quot;',"'":'&#39;'}[c]));
  }

  const state = {
    pdfDoc: null, selectedYear: null, timeLimitSec: 90*60, currentQuestion: 1,
    flagged: {1:false,2:false,3:false,4:false},
    startTimestamp: null, endTimestamp: null, timerInterval: null,
    zoom: 1.3, isPreset: false, pdfRendered: false,
    refDoc: null, refRendered: false,
  };

  // ---- Build year grid ----
  const yearGrid = $('exam-year-grid');
  PRESET_YEARS.forEach(year => {
    const btn = document.createElement('button');
    btn.className = 'exam-year-chip';
    btn.dataset.year = year;
    const mins = timeLimitMinutes(year);
    btn.innerHTML = year + '<div class="chip-sub">' + (mins === 105 ? '1h 45m' : '1h 30m') + '</div>';
    btn.addEventListener('click', () => selectPresetYear(year));
    yearGrid.appendChild(btn);
  });

  function selectPresetYear(year) {
    yearGrid.querySelectorAll('.exam-year-chip').forEach(c => c.classList.remove('selected'));
    const chip = yearGrid.querySelector('[data-year="' + year + '"]');
    if (chip) chip.classList.add('selected');

    state.selectedYear = year;
    state.isPreset = true;
    state.pdfDoc = null;
    $('exam-file-list').innerHTML = '';

    const mins = timeLimitMinutes(year);
    state.timeLimitSec = mins * 60;
    $('exam-time-value').textContent = Math.floor(mins/60) + 'h ' + (mins%60) + 'm';
    $('exam-selected-label').textContent = year + ' FRQ';
    $('exam-time-row').classList.remove('exam-hidden');
    $('exam-continue-btn').disabled = false;
  }

  // ---- File upload (custom) ----
  const dropzone = $('exam-dropzone');
  const fileInput = $('exam-file-input');
  const fileList = $('exam-file-list');

  ['dragenter','dragover'].forEach(ev => {
    dropzone.addEventListener(ev, e => { e.preventDefault(); e.stopPropagation(); dropzone.classList.add('dragover'); });
  });
  ['dragleave','drop'].forEach(ev => {
    dropzone.addEventListener(ev, e => { e.preventDefault(); e.stopPropagation(); dropzone.classList.remove('dragover'); });
  });
  dropzone.addEventListener('drop', e => {
    const pdfs = Array.from(e.dataTransfer.files).filter(f => f.type === 'application/pdf');
    if (pdfs.length) handleFile(pdfs[0]);
  });
  fileInput.addEventListener('change', e => {
    const pdfs = Array.from(e.target.files).filter(f => f.type === 'application/pdf');
    if (pdfs.length) handleFile(pdfs[0]);
  });

  async function handleFile(file) {
    yearGrid.querySelectorAll('.exam-year-chip').forEach(c => c.classList.remove('selected'));
    state.isPreset = false;
    try {
      const ab = await file.arrayBuffer();
      state.pdfDoc = await pdfjsLib.getDocument({data: ab}).promise;
      let text = '';
      for (let i = 1; i <= Math.min(2, state.pdfDoc.numPages); i++) {
        const page = await state.pdfDoc.getPage(i);
        const content = await page.getTextContent();
        text += content.items.map(item => item.str).join(' ') + ' ';
      }
      state.selectedYear = detectYear(text) || 2024;
      const mins = timeLimitMinutes(state.selectedYear);
      state.timeLimitSec = mins * 60;
      $('exam-time-value').textContent = Math.floor(mins/60) + 'h ' + (mins%60) + 'm';
      $('exam-selected-label').textContent = 'Custom PDF';
      $('exam-time-row').classList.remove('exam-hidden');
      renderUploadedFileList(file.name, state.pdfDoc.numPages);
      $('exam-continue-btn').disabled = false;
    } catch(e) { alert('Failed to load: ' + file.name); }
  }

  function detectYear(text) {
    const matches = Array.from(text.matchAll(/\b(20\d{2})\b/g)).map(m=>parseInt(m[1],10));
    const valid = matches.filter(y => y >= 2004 && y <= new Date().getFullYear()+1);
    if (!valid.length) return null;
    const counts = {};
    valid.forEach(y => counts[y]=(counts[y]||0)+1);
    return parseInt(Object.entries(counts).sort((a,b)=>b[1]-a[1])[0][0],10);
  }

  function renderUploadedFileList(name, pages) {
    fileList.innerHTML = '';
    const li = document.createElement('li');
    li.innerHTML = '<div><span style="font-weight:500;">' + escapeHtml(name) + '</span> <span class="exam-file-meta">' + pages + ' pg</span></div><button>✕</button>';
    li.querySelector('button').addEventListener('click', () => {
      state.pdfDoc = null; fileList.innerHTML = '';
      $('exam-continue-btn').disabled = true;
      $('exam-time-row').classList.add('exam-hidden');
    });
    fileList.appendChild(li);
  }

  // ---- Screen switching ----
  function showScreen(name) {
    ['upload','instructions','main','review'].forEach(s => {
      const el = $('exam-'+s+'-screen');
      if (el) el.classList.toggle('exam-hidden', s!==name);
    });
  }

  // ---- Relocate code runners into exam layout ----
  function relocateRunners() {
    for (let q=1;q<=4;q++) {
      const src=$('exam-coderunner-q'+q), tgt=$('exam-runner-q'+q);
      if (src&&tgt) { while(src.firstChild) tgt.appendChild(src.firstChild); src.style.display='none'; }
    }
  }

  // ---- Continue / Instructions ----
  $('exam-continue-btn').addEventListener('click', () => {
    $('exam-inst-time').textContent = formatTime(state.timeLimitSec);
    $('exam-inst-year').textContent = state.selectedYear;
    showScreen('instructions');
  });
  $('exam-back-to-upload').addEventListener('click', () => showScreen('upload'));
  $('exam-start-btn').addEventListener('click', beginExam);

  // ---- Load preset PDF ----
  async function loadPresetPdf() {
    const url = presetPdfUrl(state.selectedYear);
    try {
      const resp = await fetch(url);
      if (!resp.ok) throw new Error('HTTP ' + resp.status);
      const ab = await resp.arrayBuffer();
      state.pdfDoc = await pdfjsLib.getDocument({data:ab}).promise;
    } catch(e) {
      alert('Could not load preset PDF for ' + state.selectedYear + '. ' + e.message);
    }
  }

  // ---- Begin exam ----
  async function beginExam() {
    relocateRunners();
    if (state.isPreset) await loadPresetPdf();
    state.pdfRendered = false;
    showScreen('main');
    state.startTimestamp = Date.now();
    state.currentQuestion = 1;
    startTimer();
    switchQuestion(1);
    await renderFullPdf();
    // Preload reference sheet in the background
    loadRefSheetPdf();
    setTimeout(()=>{
      document.querySelectorAll('.exam-code-pane .CodeMirror').forEach(el=>{if(el.CodeMirror)el.CodeMirror.refresh();});
    },150);
  }

  // ---- Timer ----
  function startTimer() {
    updateTimer();
    state.timerInterval = setInterval(()=>{
      const rem=state.timeLimitSec-Math.floor((Date.now()-state.startTimestamp)/1000);
      if(rem<=0){clearInterval(state.timerInterval);doSubmit(true);return;}
      updateTimer(rem);
    },1000);
  }
  function updateTimer(rem) {
    if(rem===undefined) rem=state.timeLimitSec-Math.floor((Date.now()-state.startTimestamp)/1000);
    const el=$('exam-timer'); el.textContent=formatTime(rem);
    el.classList.remove('warning','danger');
    if(rem<=5*60)el.classList.add('danger'); else if(rem<=10*60)el.classList.add('warning');
  }

  // ---- PDF rendering (FRQ) ----
  async function renderFullPdf() {
    const wrap=$('exam-pdf-canvases'); wrap.innerHTML='';
    if(!state.pdfDoc) return;
    const totalPages = state.pdfDoc.numPages;
    $('exam-pdf-title').textContent = 'FRQ ' + state.selectedYear + ' — ' + totalPages + ' pages';
    for(let i=1; i<=totalPages; i++){
      const page = await state.pdfDoc.getPage(i);
      const vp = page.getViewport({scale:state.zoom});
      const c = document.createElement('canvas'); c.width=vp.width; c.height=vp.height;
      wrap.appendChild(c);
      const lbl = document.createElement('div'); lbl.className='exam-pdf-page-label';
      lbl.textContent = 'Page ' + i + ' of ' + totalPages;
      wrap.appendChild(lbl);
      await page.render({canvasContext:c.getContext('2d'),viewport:vp}).promise;
    }
    state.pdfRendered = true;
    $('exam-pdf-pane').scrollTop = 0;
  }
  $('exam-zoom-in').addEventListener('click',()=>{state.zoom=Math.min(2.5,state.zoom+0.2);renderFullPdf();});
  $('exam-zoom-out').addEventListener('click',()=>{state.zoom=Math.max(0.6,state.zoom-0.2);renderFullPdf();});

  // ---- Reference Sheet ----
  async function loadRefSheetPdf() {
    if (state.refDoc) return; // already loaded
    try {
      const resp = await fetch(refSheetUrl());
      if (!resp.ok) throw new Error('HTTP ' + resp.status);
      const ab = await resp.arrayBuffer();
      state.refDoc = await pdfjsLib.getDocument({data: ab}).promise;
    } catch(e) {
      console.warn('Could not preload reference sheet:', e.message);
    }
  }

  async function renderRefSheet() {
    const wrap = $('exam-ref-canvases');
    if (state.refRendered) return; // already rendered
    wrap.innerHTML = '<div class="exam-ref-loading">Rendering reference sheet…</div>';

    if (!state.refDoc) {
      // Try loading now if preload didn't finish yet
      try {
        const resp = await fetch(refSheetUrl());
        if (!resp.ok) throw new Error('HTTP ' + resp.status);
        const ab = await resp.arrayBuffer();
        state.refDoc = await pdfjsLib.getDocument({data: ab}).promise;
      } catch(e) {
        wrap.innerHTML = '<div class="exam-ref-loading" style="color:#fca5a5;">Could not load reference sheet.<br>' + escapeHtml(e.message) + '</div>';
        return;
      }
    }

    wrap.innerHTML = '';
    for (let i = 1; i <= state.refDoc.numPages; i++) {
      const page = await state.refDoc.getPage(i);
      // Scale to fit the 480px drawer width (leaving room for padding)
      const vp = page.getViewport({scale: 1.0});
      const scale = 460 / vp.width;
      const scaledVp = page.getViewport({scale});
      const c = document.createElement('canvas');
      c.width = scaledVp.width; c.height = scaledVp.height;
      wrap.appendChild(c);
      await page.render({canvasContext: c.getContext('2d'), viewport: scaledVp}).promise;
    }
    state.refRendered = true;
  }

  function openRefDrawer() {
    $('exam-ref-drawer').classList.add('open');
    $('exam-ref-backdrop').classList.add('open');
    $('exam-ref-btn').classList.add('active');
    renderRefSheet();
  }
  function closeRefDrawer() {
    $('exam-ref-drawer').classList.remove('open');
    $('exam-ref-backdrop').classList.remove('open');
    $('exam-ref-btn').classList.remove('active');
  }

  $('exam-ref-btn').addEventListener('click', () => {
    if ($('exam-ref-drawer').classList.contains('open')) closeRefDrawer();
    else openRefDrawer();
  });
  $('exam-ref-close').addEventListener('click', closeRefDrawer);
  $('exam-ref-backdrop').addEventListener('click', closeRefDrawer);

  // ---- Question switching ----
  function switchQuestion(q) {
    state.currentQuestion=q;
    for(let i=1;i<=4;i++){const el=$('exam-runner-q'+i);if(el)el.classList.toggle('exam-hidden',i!==q);}
    $('exam-code-label').textContent='Your Response — Question '+q;
    $('exam-footer-info').textContent='Question '+q+' of 4';
    $('exam-prev-q').disabled=q===1; $('exam-next-q').disabled=q===4;
    updateQNav(); updateFlagBtn();
    setTimeout(()=>{const r=$('exam-runner-q'+q);if(r){const cm=r.querySelector('.CodeMirror');if(cm&&cm.CodeMirror)cm.CodeMirror.refresh();}},50);
  }
  document.querySelectorAll('.exam-qnav-btn').forEach(b=>b.addEventListener('click',()=>switchQuestion(parseInt(b.dataset.q,10))));
  $('exam-prev-q').addEventListener('click',()=>{if(state.currentQuestion>1)switchQuestion(state.currentQuestion-1);});
  $('exam-next-q').addEventListener('click',()=>{if(state.currentQuestion<4)switchQuestion(state.currentQuestion+1);});

  function updateQNav(){document.querySelectorAll('.exam-qnav-btn').forEach(b=>{const q=parseInt(b.dataset.q,10);b.classList.toggle('active',q===state.currentQuestion);b.classList.toggle('flagged',state.flagged[q]);});}

  // ---- Flag ----
  $('exam-flag-btn').addEventListener('click',()=>{state.flagged[state.currentQuestion]=!state.flagged[state.currentQuestion];updateFlagBtn();updateQNav();});
  function updateFlagBtn(){const f=state.flagged[state.currentQuestion];const b=$('exam-flag-btn');b.classList.toggle('active',f);b.textContent=f?'Flagged':'Flag';}

  // ---- Exit ----
  $('exam-exit-btn').addEventListener('click',()=>{
    showModal('Exit exam?','Your responses will be discarded. Are you sure?',()=>{
      clearInterval(state.timerInterval);state.flagged={1:false,2:false,3:false,4:false};
      closeRefDrawer();
      showScreen('upload');
    });
  });

  // ---- Get code ----
  function getAnswerCode(q){const r=$('exam-runner-q'+q);if(!r)return '';const cm=r.querySelector('.CodeMirror');if(cm&&cm.CodeMirror)return cm.CodeMirror.getValue();const ta=r.querySelector('.editor-textarea');return ta?ta.value:'';}

  // ---- Submit ----
  $('exam-submit-btn').addEventListener('click',()=>{
    const u=[1,2,3,4].filter(q=>!getAnswerCode(q).trim());
    const msg=u.length?'You have '+u.length+' unanswered question(s) ('+u.join(', ')+'). Submit anyway?':'Are you sure you want to submit? This cannot be undone.';
    showModal('Submit exam?',msg,()=>doSubmit(false));
  });
  function doSubmit(timedOut){clearInterval(state.timerInterval);state.endTimestamp=Date.now();closeRefDrawer();buildReview(timedOut);showScreen('review');}

  // ---- Review ----
  function buildReview(timedOut){
    const elapsed=Math.floor((state.endTimestamp-state.startTimestamp)/1000);
    const used=Math.min(elapsed,state.timeLimitSec);
    const answers={};for(let q=1;q<=4;q++)answers[q]=getAnswerCode(q);
    const answered=[1,2,3,4].filter(q=>answers[q].trim()).length;
    $('exam-review-subtitle').textContent=timedOut?'Time expired — auto-submitted.':'Your exam has been submitted.';
    $('exam-rv-year').textContent=state.selectedYear;
    $('exam-rv-time').textContent=formatTime(used);
    $('exam-rv-allowed').textContent=formatTime(state.timeLimitSec);
    $('exam-rv-answered').textContent=answered+' / 4';
    const container=$('exam-review-questions');container.innerHTML='';
    for(let q=1;q<=4;q++){
      const code=answers[q],flag=state.flagged[q]?' (flagged)':'';
      const div=document.createElement('div');div.className='exam-q-review';
      div.innerHTML='<h3>Question '+q+flag+'<span style="font-size:11px;font-weight:500;color:var(--ex-text-muted);">'+code.trim().length+' chars</span></h3>'
        +(code.trim()?'<pre>'+escapeHtml(code)+'</pre>':'<div class="exam-empty-answer">No response submitted.</div>')
        +'<div id="exam-ai-feedback-q'+q+'" class="exam-ai-feedback exam-hidden"></div>';
      container.appendChild(div);
    }
    state._reviewAnswers=answers;state._reviewUsed=used;

    // Reset grade button
    const gradeBtn=$('exam-grade-ai-btn');
    gradeBtn.disabled=false;
    gradeBtn.innerHTML='✨ Grade with AI';
    $('exam-grading-progress').classList.add('exam-hidden');
  }

  // ---- AI Grading via Gemini ----
  $('exam-grade-ai-btn').addEventListener('click', startAIGrading);

  async function startAIGrading() {
    const btn=$('exam-grade-ai-btn');
    const progress=$('exam-grading-progress');
    const answers=state._reviewAnswers||{};

    const toGrade=[1,2,3,4].filter(q=>(answers[q]||'').trim().length>0);
    if(!toGrade.length){
      progress.classList.remove('exam-hidden');
      progress.textContent='No answers to grade.';
      return;
    }

    btn.disabled=true;
    btn.innerHTML='<span class="spinner"></span> Grading...';
    progress.classList.remove('exam-hidden');
    progress.textContent='Grading question '+toGrade[0]+' of '+toGrade.length+'...';

    let graded=0;
    for(const q of toGrade){
      progress.textContent='Grading Question '+q+'... ('+(graded+1)+' of '+toGrade.length+')';
      const fbDiv=$('exam-ai-feedback-q'+q);
      fbDiv.classList.remove('exam-hidden');
      fbDiv.innerHTML='<div class="feedback-loading">Sending to Gemini AI for grading...</div>';

      try {
        const res=await fetch(javaURI+'/api/gemini-frq/grade',{
          method:'POST', mode:'cors', credentials:'include',
          headers:{'Content-Type':'application/json'},
          body:JSON.stringify({question:'AP CSA '+state.selectedYear+' FRQ Question '+q, answer:answers[q]})
        });

        if(!res.ok) throw new Error('Server returned '+res.status);
        const data=await res.json();

        const feedback=data.feedback
          || data.gradingResult
          || (data.candidates&&data.candidates[0]&&data.candidates[0].content&&data.candidates[0].content.parts&&data.candidates[0].content.parts[0]&&data.candidates[0].content.parts[0].text)
          || 'No feedback returned.';

        const scoreMatch=feedback.match(/Grade:\s*([\d.]+)\s*\/\s*([\d.]+)/i);
        let scoreHtml='';
        if(scoreMatch) scoreHtml='<span class="feedback-score">'+scoreMatch[1]+' / '+scoreMatch[2]+'</span><br>';

        const formatted=feedback.replace(/\*\*(.*?)\*\*/g,'<strong>$1</strong>').replace(/\n/g,'<br>');
        fbDiv.innerHTML=scoreHtml+formatted;
      } catch(err){
        fbDiv.innerHTML='<div class="feedback-error">Grading failed: '+escapeHtml(err.message)+'. Make sure you are logged in and the backend is running.</div>';
      }
      graded++;
    }

    btn.innerHTML='✅ Grading Complete';
    progress.textContent='All '+toGrade.length+' question(s) graded.';
  }

  // ---- Download ----
  $('exam-download-btn').addEventListener('click',()=>{
    const a=state._reviewAnswers||{},u=state._reviewUsed||0;
    let t='AP Computer Science A — Section II\nExam year: '+state.selectedYear+'\nTime allowed: '+formatTime(state.timeLimitSec)+'\nTime used: '+formatTime(u)+'\nDate: '+new Date().toLocaleString()+'\n'+'='.repeat(60)+'\n\n';
    for(let q=1;q<=4;q++){t+='QUESTION '+q+(state.flagged[q]?' (flagged)':'')+'\n'+'-'.repeat(60)+'\n'+(a[q]||'[no response]')+'\n\n';}
    const blob=new Blob([t],{type:'text/plain'}),url=URL.createObjectURL(blob),el=document.createElement('a');
    el.href=url;el.download='apcsa-responses-'+state.selectedYear+'.txt';el.click();URL.revokeObjectURL(url);
  });

  // ---- New exam ----
  $('exam-new-btn').addEventListener('click',()=>{
    state.pdfDoc=null;state.selectedYear=null;state.timeLimitSec=90*60;state.isPreset=false;
    state.pdfRendered=false;state.flagged={1:false,2:false,3:false,4:false};
    $('exam-file-list').innerHTML='';
    yearGrid.querySelectorAll('.exam-year-chip').forEach(c=>c.classList.remove('selected'));
    $('exam-time-row').classList.add('exam-hidden');
    $('exam-continue-btn').disabled=true;
    showScreen('upload');
  });

  // ---- Modal ----
  function showModal(title,body,onConfirm){
    $('exam-modal-title').textContent=title;$('exam-modal-body').textContent=body;
    $('exam-modal').classList.remove('exam-hidden');
    const close=()=>$('exam-modal').classList.add('exam-hidden');
    const ch=()=>{close();cleanup();onConfirm();};
    const cc=()=>{close();cleanup();};
    function cleanup(){$('exam-modal-confirm').removeEventListener('click',ch);$('exam-modal-cancel').removeEventListener('click',cc);}
    $('exam-modal-confirm').addEventListener('click',ch);
    $('exam-modal-cancel').addEventListener('click',cc);
  }

  window.addEventListener('beforeunload',e=>{if(!$('exam-main-screen').classList.contains('exam-hidden')){e.preventDefault();e.returnValue='';}});
})();
