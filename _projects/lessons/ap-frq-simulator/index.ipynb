    gradeBtn.disabled=false;
    gradeBtn.innerHTML='✨ Grade with AI';
    $('exam-grading-progress').classList.add('exam-hidden');
  }

  // ---- AI Grading via Gemini ----
  $('exam-grade-ai-btn').addEventListener('click', startAIGrading);

  async function startAIGrading() {
    const btn=$('exam-grade-ai-btn');
    const progress=$('exam-grading-progress');
    const answers=state._reviewAnswers||{};

    const toGrade=[1,2,3,4].filter(q=>(answers[q]||'').trim().length>0);
    if(!toGrade.length){
      progress.classList.remove('exam-hidden');
      progress.textContent='No answers to grade.';
      return;
    }

    btn.disabled=true;
    btn.innerHTML='<span class="spinner"></span> Grading...';
    progress.classList.remove('exam-hidden');
    progress.textContent='Grading question '+toGrade[0]+' of '+toGrade.length+'...';

    let graded=0;
    for(const q of toGrade){
      progress.textContent='Grading Question '+q+'... ('+(graded+1)+' of '+toGrade.length+')';
      const fbDiv=$('exam-ai-feedback-q'+q);
      fbDiv.classList.remove('exam-hidden');
      fbDiv.innerHTML='<div class="feedback-loading">Sending to Gemini AI for grading...</div>';

      try {
        const res=await fetch(javaURI+'/api/gemini-frq/grade',{
          method:'POST', mode:'cors', credentials:'include',
          headers:{'Content-Type':'application/json'},
          body:JSON.stringify({question:'AP CSA '+state.selectedYear+' FRQ Question '+q, answer:answers[q]})
        });

        if(!res.ok) throw new Error('Server returned '+res.status);
        const data=await res.json();

        const feedback=data.feedback
          || data.gradingResult
          || (data.candidates&&data.candidates[0]&&data.candidates[0].content&&data.candidates[0].content.parts&&data.candidates[0].content.parts[0]&&data.candidates[0].content.parts[0].text)
          || 'No feedback returned.';

        const scoreMatch=feedback.match(/Grade:\s*([\d.]+)\s*\/\s*([\d.]+)/i);
        let scoreHtml='';
        if(scoreMatch) scoreHtml='<span class="feedback-score">'+scoreMatch[1]+' / '+scoreMatch[2]+'</span><br>';

        const formatted=feedback.replace(/\*\*(.*?)\*\*/g,'<strong>$1</strong>').replace(/\n/g,'<br>');
        fbDiv.innerHTML=scoreHtml+formatted;
      } catch(err){
        fbDiv.innerHTML='<div class="feedback-error">Grading failed: '+escapeHtml(err.message)+'. Make sure you are logged in and the backend is running.</div>';
      }
      graded++;
    }

    btn.innerHTML='✅ Grading Complete';
    progress.textContent='All '+toGrade.length+' question(s) graded.';
  }
