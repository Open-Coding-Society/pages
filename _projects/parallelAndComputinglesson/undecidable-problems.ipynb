<script>
(function() {
  const programs = [
    {
      label: 'PROGRAM 1 OF 4',
      desc: 'A loop that counts down from 5.',
      answer: 'stops',
      lines: [
        { text: 'x = 5', cls: 'num' },
        { text: 'while x > 0:', cls: 'kw' },
        { text: '    x = x - 1', cls: '' },
        { text: 'print("done")', cls: 'fn' },
      ],
      runLines: [0,1,2,1,2,1,2,1,2,1,3],
      correct: '✅ Correct — it stops.',
      wrong: '❌ Not quite — it stops.',
      explain: 'x starts at 5 and shrinks by 1 each loop. After 5 loops it hits 0, the condition fails, and the program ends.',
    },
    {
      label: 'PROGRAM 2 OF 4',
      desc: 'A loop that keeps growing.',
      answer: 'loops',
      lines: [
        { text: 'x = 1', cls: 'num' },
        { text: 'while x > 0:', cls: 'kw' },
        { text: '    x = x + 1', cls: '' },
        { text: 'print("done")', cls: 'fn' },
      ],
      runLines: [0,1,2,1,2,1,2,1,2,1,2],
      correct: '✅ Correct — it loops forever.',
      wrong: '❌ Not quite — it loops forever.',
      explain: 'x starts at 1 and only ever increases. The condition x > 0 will never be false, so the loop never ends.',
    },
    {
      label: 'PROGRAM 3 OF 4',
      desc: 'The Collatz conjecture — a real unsolved math problem. Take any number. If even, halve it. If odd, multiply by 3 and add 1. Does it always reach 1?',
      answer: 'stops',
      lines: [
        { text: 'n = 27', cls: 'num' },
        { text: 'while n != 1:', cls: 'kw' },
        { text: '    if n % 2 == 0:', cls: 'kw' },
        { text: '        n = n // 2', cls: '' },
        { text: '    else:', cls: 'kw' },
        { text: '        n = 3 * n + 1', cls: '' },
        { text: 'print("reached 1")', cls: 'fn' },
      ],
      runLines: [0,1,2,5,1,2,3,1,2,5,1,2,3,1,6],
      correct: '✅ For n=27 it stops — after 111 steps.',
      wrong: '❌ For n=27 it actually stops — after 111 steps.',
      explain: 'This sequence has been tested for every number up to 2^68 and always reaches 1. But nobody has proven it works for ALL numbers. That\'s an open problem in mathematics — meaning even experts can\'t always predict it.',
    },
    {
      label: 'PROGRAM 4 OF 4 — the trap',
      desc: '⚠️ This program reads your prediction and does the opposite of whatever you pick.',
      answer: '__rigged__',
      lines: [
        { text: '# your_guess = whatever you just clicked', cls: 'comment' },
        { text: '', cls: '' },
        { text: 'def trouble(your_guess):', cls: 'fn' },
        { text: '    if your_guess == "stops":', cls: 'kw' },
        { text: '        while True: pass  # loop forever', cls: 'kw' },
        { text: '    else:', cls: 'kw' },
        { text: '        return           # stop immediately', cls: '' },
        { text: '', cls: '' },
        { text: 'trouble(your_guess)', cls: 'fn' },
      ],
      runLines: [0,2,3,4,4,4,3,4,4,4],
    },
  ];

  let idx = 0, correct = 0, total = 0, answered = false, runInterval = null;

  function render() {
    const p = programs[idx];
    document.getElementById('hg-label').textContent = p.label;
    document.getElementById('hg-desc').textContent = p.desc;
    document.getElementById('hg-result').style.display = 'none';
    document.getElementById('hg-btns').style.display = 'flex';
    document.getElementById('hg-next-btn').style.display = 'none';
    document.getElementById('hg-running').style.display = 'none';
    document.getElementById('hg-final').style.display = 'none';
    document.getElementById('hg-code').style.display = 'block';
    renderCode(p.lines, -1);
    answered = false;
    updateScore();
  }

  function renderCode(lines, activeLine) {
    document.getElementById('hg-code').innerHTML = lines.map(function(l, i) {
      return '<span class="hg-line ' + l.cls + (i === activeLine ? ' active' : '') + '">' + (l.text || '&nbsp;') + '</span>';
    }).join('');
  }

  function startRun(p, guess) {
    var el = document.getElementById('hg-running');
    el.style.display = 'block';
    var step = 0, dots = 0;
    var ticker = document.getElementById('hg-ticker');
    var extraSteps = idx === 3 ? 8 : 0;
    runInterval = setInterval(function() {
      dots = (dots + 1) % 4;
      ticker.textContent = '▶ Running' + '.'.repeat(dots);
      renderCode(p.lines, p.runLines[step % p.runLines.length]);
      step++;
      if (step >= p.runLines.length + extraSteps) {
        clearInterval(runInterval);
        el.style.display = 'none';
        renderCode(p.lines, -1);
        showResult(p, guess);
      }
    }, idx === 3 ? 160 : 120);
  }

  function showResult(p, guess) {
    if (p.answer === '__rigged__') { showTrap(guess); return; }
    var won = guess === p.answer;
    if (won) correct++;
    total++;
    var r = document.getElementById('hg-result');
    r.style.display = 'block';
    r.style.background = won ? '#0d2a0d' : '#2a0808';
    r.style.borderColor = won ? '#4caf50' : '#ef5350';
    r.style.color = won ? '#c8e6c9' : '#ffcdd2';
    r.innerHTML = '<strong>' + (won ? p.correct : p.wrong) + '</strong><br>' + p.explain;
    document.getElementById('hg-next-btn').style.display = 'inline-block';
    updateScore();
  }

  function showTrap(guess) {
    total++;
    var opposite = guess === 'stops' ? 'loop forever' : 'stop immediately';
    var r = document.getElementById('hg-result');
    r.style.display = 'block';
    r.style.background = '#1a0808';
    r.style.borderColor = '#cc0022';
    r.style.color = '#ffcdd2';
    r.innerHTML = '<strong>❌ Wrong — it chose to ' + opposite + '.</strong><br>Whatever you predicted, this program did the opposite. You cannot win this one — and that\'s the whole point.';
    document.getElementById('hg-next-btn').textContent = 'See why this matters →';
    document.getElementById('hg-next-btn').style.display = 'inline-block';
    updateScore();
  }

  function updateScore() {
    document.getElementById('hg-score').textContent = total > 0 ? 'Accuracy: ' + correct + '/' + total : 'Accuracy: —';
  }

  window.hgGuess = function(guess) {
    if (answered) return;
    answered = true;
    document.getElementById('hg-btns').style.display = 'none';
    startRun(programs[idx], guess);
  };

  window.hgNext = function() {
    idx++;
    if (idx >= programs.length) { showFinal(); } else { render(); }
  };

  function showFinal() {
    document.getElementById('hg-code').style.display = 'none';
    document.getElementById('hg-result').style.display = 'none';
    document.getElementById('hg-next-btn').style.display = 'none';
    document.getElementById('hg-final').style.display = 'block';
    document.getElementById('hg-label').textContent = 'RESULT';
    document.getElementById('hg-desc').textContent = '';
    document.getElementById('hg-final-msg').innerHTML = 'You got ' + correct + '/3 — but program 4 was impossible to get right.';
    document.getElementById('hg-final-sub').innerHTML =
      'That last program is the core of the <strong style="color:#fff">Halting Problem</strong>. Alan Turing proved in 1936 that no algorithm can predict whether any arbitrary program will stop — because you can always build one that does the opposite of your prediction. Any problem that requires solving this is also impossible to solve in general. Including surviving Tetris forever.';
  }

  render();
})();
</script>
