```python
%%js

// ═══════════════════════════════════════════════════════════════
// GAME_RUNNER: AI Wizard Dialogue System (Replaces Quiz System)
// ═══════════════════════════════════════════════════════════════

// Imports
import { GameControl, GameEnvBackground, Player, NPC, AiNpc } 
from '/assets/js/GameEnginev1.1/essentials/Imports.js';

// ═══════════════════════════════════════════════════════════════
// LEVEL CLASS
// ═══════════════════════════════════════════════════════════════

class CustomLevel {
  constructor(gameEnv) {
    const path   = gameEnv.path;
    const width  = gameEnv.innerWidth;
    const height = gameEnv.innerHeight;

    // ───────────────── BACKGROUND ─────────────────
    const bgData = {
      name: 'osi_bg',
      src: path + "/images/gamify/comics/network-stack/7-layers-of-osi-model.png",
    };

    // ───────────────── PLAYER ─────────────────
    const playerData = {
      id: 'Hero',
      src: path + "/images/gamify/chillguy.png",

      SCALE_FACTOR: 5,
      STEP_FACTOR: 1000,
      ANIMATION_RATE: 50,

      INIT_POSITION: { x: 0, y: height * 0.15 },

      pixels: { height: 512, width: 384 },
      orientation: { rows: 4, columns: 3 },

      down:      { row: 0, start: 0, columns: 3 },
      right:     { row: 1, start: 0, columns: 3 },
      left:      { row: 2, start: 0, columns: 3 },
      up:        { row: 3, start: 0, columns: 3 },

      hitbox: { widthPercentage: 0.45, heightPercentage: 0.2 },

      keypress: { up: 87, left: 65, down: 83, right: 68 }
    };

    // ═══════════════════════════════════════════
    // LESSON NPC TEMPLATE (REUSABLE)
    // ═══════════════════════════════════════════

    function createLessonNPC(id, text, xPos) {
      return {
        id: id,
        greeting: text,

        SCALE_FACTOR: 8,
        visible: false,

        INIT_POSITION: { x: width * xPos, y: height * 0.46 },

        hitbox: { widthPercentage: 0, heightPercentage: 0 },

        reaction: function() {
          this.showDialogue(this.greeting);
        },

        interact: function() {
          this.showDialogue(this.greeting);
        }
      };
    }

    // ───────────────── OSI NPCs ─────────────────
    const osiNPCs = [
      createLessonNPC("Application Layer",
        "Layer 7 (Application): Interfaces directly with user applications like browsers.",
        0.06),

      createLessonNPC("Presentation Layer",
        "Layer 6 (Presentation): Handles encryption, compression, and translation.",
        0.20),

      createLessonNPC("Session Layer",
        "Layer 5 (Session): Manages connections between systems.",
        0.34),

      createLessonNPC("Transport Layer",
        "Layer 4 (Transport): Ensures reliable data transfer.",
        0.48),

      createLessonNPC("Network Layer",
        "Layer 3 (Network): Handles routing and logical addressing.",
        0.62),

      createLessonNPC("Data Link Layer",
        "Layer 2 (Data Link): Node-to-node transfer and error detection.",
        0.76),

      createLessonNPC("Physical Layer",
        "Layer 1 (Physical): Transmits raw bits over hardware.",
        0.90)
    ];

    // ═══════════════════════════════════════════
    // 🧙 AI WIZARD SYSTEM (CORE FEATURE)
    // ═══════════════════════════════════════════

    const npcWizard = {
      id: "Network Wizard",

      src: path + "/images/gamify/wizard.png",

      SCALE_FACTOR: 7,
      ANIMATION_RATE: 10,

      pixels: { height: 185, width: 163 },
      INIT_POSITION: { x: width * 0.47, y: 0 },

      orientation: { rows: 1, columns: 1 },
      down: { row: 0, start: 0, columns: 1 },

      hitbox: { widthPercentage: 0.2, heightPercentage: 0.2 },

      // ───────── AI CONFIG ─────────
      expertise: "networking",

      chatHistory: [],

      knowledgeBase: {
        networking: [
          {
            keywords: ["osi", "layers"],
            response: "The OSI model has 7 layers: Physical → Application. Each layer has a specific role in communication."
          },
          {
            keywords: ["http", "https"],
            response: "HTTPS is secure because it encrypts data using TLS, while HTTP does not."
          },
          {
            keywords: ["packet"],
            response: "A packet is a unit of data that travels across a network with headers and payload."
          },
          {
            keywords: ["dns"],
            response: "DNS translates domain names into IP addresses so browsers can locate servers."
          }
        ]
      },

      // ───────── AI RESPONSE ENGINE ─────────
      generateResponse: function(input) {
        input = input.toLowerCase();

        const kb = this.knowledgeBase[this.expertise];

        for (let entry of kb) {
          for (let keyword of entry.keywords) {
            if (input.includes(keyword)) {
              return entry.response;
            }
          }
        }

        return "Interesting question... try asking about networking concepts like OSI, HTTP, or packets.";
      },

      // ───────── INTERACTION ─────────
      interact: function() {
        AiNpc.showInteraction(this);
      },

      reaction: function() {
        this.showDialogue("Greetings traveler... ask me anything about networking.");
      }
    };

    // ═══════════════════════════════════════════
    // CLASS REGISTRATION
    // ═══════════════════════════════════════════

    this.classes = [
      { class: GameEnvBackground, data: bgData },
      { class: Player, data: playerData },

      ...osiNPCs.map(npc => ({ class: NPC, data: npc })),

      { class: NPC, data: npcWizard }
    ];
  }
}

// ───────────────── EXPORTS ─────────────────
export const gameLevelClasses = [CustomLevel];
export { GameControl };
```
